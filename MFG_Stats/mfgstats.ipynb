## <div class="alert alert-warning"> Gage Repeatability and Reproducibility (GRR)
- Measurement Systems Analysis And Build Yield Mearement

#### 1. Measurement System Analysis(MSA)
- Calculate, Analyze, and interpret measurment system capability using GR&R studies, measurment correlations, bias, linearity, percent agreement, and precision/tolerance (P/T)
- General Steps:
  - Define The Problem
  - Measure The Problem and Process
  - Analyze Data To Determine The Root Cause
  - Improve The Process
  - Control The Process and Product


#### 2. Terms and Definitions
- What is Measurment System Analysis(MSA)?
  - Measurement systems have variations just like every other process.
  - All Measurments have small or big systematic or random variations.
  - Variations:
    - Process variation
    - Measurement Variation:GRR is performed to quantify this variation.
    - For a measurement system(or tester) to be capable, the measurment system variation must be small relative to th design tolerance ($\pm$ or the process variation.
- The Average and Range Method
  - Repeatability: $Equipment\ Variation(EV)\ =\ \sigma_{Repeatability}\ =\ \frac{\bar{\bar{R}}}{d_2}$
  - Reproducibility: $Appraiser\ Variation(AV)\ =\ \sigma_{Reproducibility}\ =\ \sqrt{(\frac{R_o}{d_2})^2 -\frac{EV^2}{n\times r}}$
  - Gauge R & R: $Measurement\ System\ Variation(GRR) =\ \sigma_{GRR}\ =\ \sqrt{EV^2 + AV^2}$
    - OPT/OST or APT/AST: $Operator/Appraiser||Parts/Samples||Trials\ 3x10x3$
  - Part To Part Variation: $PV =\ \sigma_{p}\ =\ \frac{R_p}{d_2}$
  - Total Variation: $TV =\ \sigma_{p}\ =\ \sqrt{GRR^2 + PV^2}$

#### 3. The Gauge R & R Process
- The purpose of GRR is to confirm that a measurment system is capable for its intended use.
- There are two methods for assessing acceptability or capability of a measurement system.
  - A. **The Precision/Tolerance Ratio (%Tolerance)**: Used if the measurement system is meant for ***product control*** (testing of product for coformance to a specification).
    - Compare measurments against specification limits.
   
  - B. **The Percent(%) of Total Process Variation (% Study Variation)**: used if the meaurement system is meant for ***process control(SPC)***, where the meaasurment variation is compared against total process variation.
     - All Measurements are put into a control chart to evaluate and control the process.
  - Both methods translate your measurment variation into a percentage value for evaluation.
 
  - Pick the right method before you start the evaluation!
 
##### 3.1 The Precision/Tolerance Ratio (%Tolerance)
- For a measurement system to be capable of meeting your needs, the normal variations (error) associated with the measurement system must be small compared to the design tolerance of the feature being measured.
- Evaluates Conformance to specification.
$$Precision\ Tolerance\ Ratio\ =\ \frac{Precision}{Tolerance}\ =\ \frac{Total\ Measurement\ System\ Variation}{Product\ Design\ Tolerance}$$
$$Precision\ Tolerance\ Ratio\ =\  \frac{6\times \sigma_{GRR}}{Upper\ Spec(USL) - Lower\ Spec(LSL)}$$
- $6\sigma (\pm 3\sigma)$ represents a $99.73\%$ spread, which is used to represent the "full' spread of measurment variation (5.15 = 99%).

- ***Interpretation***:  The P/T Ratio is the portion (percentage) of the tolerance range that is consumed by the measurment variation exclusively.
##### 3.2 The Percent(%) of Total Process Variation (% Study Variation), or Part To Part Variation
- For the measurment su=ystem to be capable of meeting your needs, the normal variation (error) associated with the measurment system must be small compared tot he totalmeasured variation of the feature being measured.
  - No need to do the full spread; only $1 \sigma$ will do!
$$Percent\ Variation\ =\ \frac{Measurment\ System\ Varition(GRR)}{Total\ Measurment\ Variation}\ =\ \frac{1\sigma}{1\sigma}$$
- Interpretation:
  - Measures What percentage of the variation came from the measurement and what percentage came from the process (part by part variation) variation.
  - $Percent\ Variation\ =\ \frac{0.5496}{3.0802}\times 100\%\ =\ 17.8\%$ came from the measurement system and the rest from the process variations.
  - The percent of variation reflects the portion (percentage) of the total variation that is consumed by the measurment system variation.

#### 4. Interpreting Results
- For a measurement system to be capable of meeting your needs, the normal variation (error) associated with the measurement system must be small compared to the total measured variation of the feature being measured.
- Generally, if the measurment system variation (error) is $<10\%$, then the measurment system is considered acceptable or capable.
- If the measurement variation is between $10\%$ and $30\%$, then the measurement system is generally considered acceptable; however, consideration should be given for risk, application, and cost. ***Grey Area***.
- If the measurment variation is greater than 30%, then the measurment system is usually not considered acceeptable.

### Gage Repeatability and Reproducibility (GRR) Pythonic Implementation

```python
class GageRR:
    
    def gage_rr(data, operator_col='Operator', part_col='Part', measurement_col='Measurement', 
                alpha=0.05, tolerance=None, print_results=True, plot_results=True):
        """
        Perform Gage R&R (Repeatability and Reproducibility) analysis.
        
        Parameters:
        -----------
        data : pandas.DataFrame
            Input data containing operators, parts, and measurements
        operator_col : str
            Column name for operators/appraisers
        part_col : str
            Column name for parts
        measurement_col : str
            Column name for measurements
        alpha : float
            Significance level for ANOVA (default 0.05)
        tolerance : float or None
            If provided, will calculate % of tolerance metrics
        print_results : bool
            Whether to print the results summary
        plot_results : bool
            Whether to generate plots
            
        Returns:
        --------
        dict
            Dictionary containing all Gage R&R metrics and components
        """
        
        # Prepare data
        operators = data[operator_col].astype('category')
        parts = data[part_col].astype('category')
        measurements = data[measurement_col]
        
        n_operators = len(operators.cat.categories)
        n_parts = len(parts.cat.categories)
        n_replicates = len(data) / (n_operators * n_parts)
        
        # Calculate basic statistics
        grand_mean = measurements.mean()
        part_means = data.groupby(part_col)[measurement_col].mean()
        operator_means = data.groupby(operator_col)[measurement_col].mean()
        
        # Calculate ranges
        range_df = data.groupby([operator_col, part_col])[measurement_col].agg(lambda x: x.max() - x.min())
        range_df = range_df.reset_index()
        range_df = range_df.rename(columns={measurement_col: 'Range'})
        
        # Calculate average range (for EV calculation)
        R_bar = range_df['Range'].mean()
        d2 = GageRR.get_d2_value(n_replicates)
        K1 = 1 / d2
        
        # Calculate operator ranges (for AV calculation)
        operator_range = operator_means.max() - operator_means.min()
        d2_star = GageRR.get_d2_star_value(n_operators)
        K2 = 1 / d2_star
        
        # Perform ANOVA
        model = ols(f'{measurement_col} ~ C({part_col}) + C({operator_col}) + C({part_col}):C({operator_col})', 
                    data=data).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)
        
        # Calculate variance components
        MS_parts = anova_table['sum_sq'][f'C({part_col})']
        MS_operators = anova_table['sum_sq'][f'C({operator_col})']
        MS_interaction = anova_table['sum_sq'][f'C({part_col}):C({operator_col})']
        MS_repeatability = anova_table['sum_sq']['Residual'] if 'Residual' in anova_table.index else MS_interaction
        
        # Calculate variance components using ANOVA method
        var_repeatability = MS_repeatability
        var_reproducibility = max(0, (MS_operators - MS_interaction) / (n_parts * n_replicates))
        var_interaction = max(0, (MS_interaction - MS_repeatability) / n_replicates)
        
        var_operators = max(0, var_reproducibility)
        var_parts = max(0, (MS_parts - MS_interaction) / (n_operators * n_replicates))
        
        var_total = var_repeatability + var_reproducibility + var_interaction + var_parts
        var_gage_rr = var_repeatability + var_reproducibility + var_interaction
        
        # Calculate standard deviations
        sd_repeatability = np.sqrt(var_repeatability)
        sd_reproducibility = np.sqrt(var_reproducibility)
        sd_interaction = np.sqrt(var_interaction) if var_interaction > 0 else 0
        sd_operators = np.sqrt(var_operators)
        sd_parts = np.sqrt(var_parts)
        sd_total = np.sqrt(var_total)
        sd_gage_rr = np.sqrt(var_gage_rr)
        
        # Calculate study variation (6 sigma)
        study_var_repeatability = 6 * sd_repeatability
        study_var_reproducibility = 6 * sd_reproducibility
        study_var_interaction = 6 * sd_interaction
        study_var_operators = 6 * sd_operators
        study_var_parts = 6 * sd_parts
        study_var_total = 6 * sd_total
        study_var_gage_rr = 6 * sd_gage_rr
        
        # Calculate percent contributions
        pc_repeatability = 100 * var_repeatability / var_total
        pc_reproducibility = 100 * var_reproducibility / var_total
        pc_interaction = 100 * var_interaction / var_total
        pc_operators = 100 * var_operators / var_total
        pc_parts = 100 * var_parts / var_total
        pc_gage_rr = 100 * var_gage_rr / var_total
        
        # Calculate ndc (number of distinct categories)
        ndc = max(1, int(1.41 * (var_parts / var_gage_rr)**0.5))
        
        # Calculate EV, AV, GRR using average and range method
        EV = R_bar * K1
        AV = np.sqrt((operator_range * K2)**2 - (EV**2 / (n_parts * n_replicates)))
        GRR = np.sqrt(EV**2 + AV**2)
        PV = (part_means.max() - part_means.min()) * GageRR.get_k3_value(n_parts)
        TV = np.sqrt(GRR**2 + PV**2)
        
        # Calculate %GRR and %PV
        percent_GRR = 100 * GRR / TV
        percent_EV = 100 * EV / TV
        percent_AV = 100 * AV / TV
        percent_PV = 100 * PV / TV
    
        # Calculate %Tolerance if tolerance is provided
        if tolerance is not None:
            percent_tol_GRR = 100 * GRR / tolerance
            percent_tol_EV = 100 * EV / tolerance
            percent_tol_AV = 100 * AV / tolerance
        else:
            percent_tol_GRR = percent_tol_EV = percent_tol_AV = None
        
        # Prepare results dictionary
        results = {
            'ANOVA Table': anova_table,
            'Variance Components': {
                'Repeatability (EV)': var_repeatability,
                'Reproducibility (AV)': var_reproducibility,
                'Interaction': var_interaction,
                'Operators': var_operators,
                'Parts': var_parts,
                'Total': var_total,
                'Gage R&R': var_gage_rr
            },
            'Standard Deviations': {
                'Repeatability (EV)': sd_repeatability,
                'Reproducibility (AV)': sd_reproducibility,
                'Interaction': sd_interaction,
                'Operators': sd_operators,
                'Parts': sd_parts,
                'Total': sd_total,
                'Gage R&R': sd_gage_rr
            },
            'Study Variation (6*SD)': {
                'Repeatability (EV)': study_var_repeatability,
                'Reproducibility (AV)': study_var_reproducibility,
                'Interaction': study_var_interaction,
                'Operators': study_var_operators,
                'Parts': study_var_parts,
                'Total': study_var_total,
                'Gage R&R': study_var_gage_rr
            },
            'Percent Contribution': {
                'Repeatability (EV)': pc_repeatability,
                'Reproducibility (AV)': pc_reproducibility,
                'Interaction': pc_interaction,
                'Operators': pc_operators,
                'Parts': pc_parts,
                'Gage R&R': pc_gage_rr
            },
            'Average and Range Method': {
                'EV': EV,
                'AV': AV,
                'GRR': GRR,
                'PV': PV,
                'TV': TV,
                '%EV': percent_EV,
                '%AV': percent_AV,
                '%GRR': percent_GRR,
                '%PV': percent_PV,
                '%Tolerance GRR': percent_tol_GRR,
                '%Tolerance EV': percent_tol_EV,
                '%Tolerance AV': percent_tol_AV,
            },
            'Number of Distinct Categories': ndc,
            'Data Summary': {
                'Number of Operators': n_operators,
                'Number of Parts': n_parts,
                'Number of Replicates': n_replicates,
                'Grand Mean': grand_mean
            }
        }
        
        if print_results:
            GageRR.print_gage_rr_results(results, tolerance)
            
        if plot_results:
            GageRR.plot_gage_rr_charts(data, operator_col, part_col, measurement_col, results)
        
        return results
    
    def get_d2_value(m):
        """Get d2 value for bias correction factor based on number of replicates"""
        d2_table = {
            2: 1.128,
            3: 1.693,
            4: 2.059,
            5: 2.326,
            6: 2.534,
            7: 2.704,
            8: 2.847,
            9: 2.970,
            10: 3.078
        }
        return d2_table.get(m, 1.0)  # Default to 1.0 if m not in table
    
    def get_d2_star_value(g):
        """Get d2* value for bias correction factor based on number of operators"""
        d2star_table = {
            2: 1.41,
            3: 1.91,
            4: 2.24,
            5: 2.48,
            6: 2.67,
            7: 2.83,
            8: 2.96,
            9: 3.08,
        10: 3.18
        }
        return d2star_table.get(g, 1.0)  # Default to 1.0 if g not in table

    def get_k3_value(n):
        """Get K3 constant based on number of parts"""
        k3_table = {
            2: 0.7071,
            3: 0.5231,
            4: 0.4467,
            5: 0.4030,
            6: 0.3742,
            7: 0.3534,
            8: 0.3375,
            9: 0.3249,
            10: 0.3146
        }
        return k3_table.get(n, 1.0)  # Default to 1.0 if n not in table

    def print_gage_rr_results(results, tolerance=None):
        """Print formatted Gage R&R results"""
        print("\n" + "="*80)
        print("GAGE R&R (REPEATABILITY AND REPRODUCIBILITY) ANALYSIS")
        print("="*80)
        
        # Print data summary
        print("\nDATA SUMMARY:")
        print(f"Number of Operators: {results['Data Summary']['Number of Operators']}")
        print(f"Number of Parts: {results['Data Summary']['Number of Parts']}")
        print(f"Number of Replicates: {results['Data Summary']['Number of Replicates']}")
        print(f"Grand Mean: {results['Data Summary']['Grand Mean']:.4f}")
    
        # Print ANOVA table
        print("\nANOVA TABLE:")
        print(results['ANOVA Table'])
        
        # Print variance components
        print("\nVARIANCE COMPONENTS:")
        for name, value in results['Variance Components'].items():
            print(f"{name:<25}: {value:.6f}")
        
        # Print standard deviations
        print("\nSTANDARD DEVIATIONS:")
        for name, value in results['Standard Deviations'].items():
            print(f"{name:<25}: {value:.6f}")
        
        # Print study variation
        print("\nSTUDY VARIATION (6 * SD):")
        for name, value in results['Study Variation (6*SD)'].items():
            print(f"{name:<25}: {value:.6f}")
        
        # Print percent contribution
        print("\nPERCENT CONTRIBUTION:")
        for name, value in results['Percent Contribution'].items():
            print(f"{name:<25}: {value:.2f}%")
    
        # Print average and range method results
        print("\nAVERAGE AND RANGE METHOD RESULTS:")
        print(f"{'Repeatability (EV)':<25}: {results['Average and Range Method']['EV']:.6f}")
        print(f"{'Reproducibility (AV)':<25}: {results['Average and Range Method']['AV']:.6f}")
        print(f"{'Gage R&R (GRR)':<25}: {results['Average and Range Method']['GRR']:.6f}")
        print(f"{'Part Variation (PV)':<25}: {results['Average and Range Method']['PV']:.6f}")
        print(f"{'Total Variation (TV)':<25}: {results['Average and Range Method']['TV']:.6f}")
        print(f"{'%EV':<25}: {results['Average and Range Method']['%EV']:.2f}%")
        print(f"{'%AV':<25}: {results['Average and Range Method']['%AV']:.2f}%")
        print(f"{'%GRR':<25}: {results['Average and Range Method']['%GRR']:.2f}%")
        print(f"{'%PV':<25}: {results['Average and Range Method']['%PV']:.2f}%")
        
        if tolerance is not None:
            print("\nPERCENT OF TOLERANCE:")
            print(f"{'%Tolerance (EV)':<25}: {results['Average and Range Method']['%Tolerance EV']:.2f}%")
            print(f"{'%Tolerance (AV)':<25}: {results['Average and Range Method']['%Tolerance AV']:.2f}%")
            print(f"{'%Tolerance (GRR)':<25}: {results['Average and Range Method']['%Tolerance GRR']:.2f}%")
        
        # Print number of distinct categories
        print(f"\nNUMBER OF DISTINCT CATEGORIES (ndc): {results['Number of Distinct Categories']}")
    
        # Print interpretation
        print("\nINTERPRETATION:")
        grr_percent = results['Average and Range Method']['%GRR']
        if grr_percent <= 10:
            print("The measurement system is EXCELLENT (<= 10% of total variation)")
        elif grr_percent <= 30:
            print("The measurement system is ACCEPTABLE (<= 30% of total variation)")
        else:
            print("The measurement system NEEDS IMPROVEMENT (> 30% of total variation)")
    
        ndc = results['Number of Distinct Categories']
        if ndc >= 5:
            print(f"The measurement system can distinguish {ndc} categories - GOOD")
        elif ndc >= 2:
            print(f"The measurement system can distinguish {ndc} categories - MARGINAL")
        else:
            print("The measurement system cannot distinguish part variation - NEEDS IMPROVEMENT")
        
        print("="*80 + "\n")

    def plot_gage_rr_charts(data, operator_col, part_col, measurement_col, results):
        """Generate Gage R&R charts"""
        plt.figure(figsize=(15, 10))
        
        # X-bar chart by operator
        plt.subplot(2, 2, 1)
        operator_means = data.groupby([operator_col, part_col])[measurement_col].mean().unstack()
        for operator in operator_means.index:
            plt.plot(operator_means.columns, operator_means.loc[operator], 'o-', label=f'Operator {operator}')
        plt.axhline(results['Data Summary']['Grand Mean'], color='k', linestyle='--')
        plt.title('X-bar Chart by Operator')
        plt.xlabel('Part')
        plt.ylabel('Mean Measurement')
        plt.legend()
        plt.grid()
    
        # R chart
        plt.subplot(2, 2, 2)
        range_df = data.groupby([operator_col, part_col])[measurement_col].agg(lambda x: x.max() - x.min())
        range_df = range_df.unstack()
        for operator in range_df.index:
            plt.plot(range_df.columns, range_df.loc[operator], 'o-', label=f'Operator {operator}')
        plt.axhline(range_df.values.mean(), color='k', linestyle='--')
        plt.title('Range Chart')
        plt.xlabel('Part')
        plt.ylabel('Range')
        plt.legend()
        plt.grid()
        
        # Components of variation
        plt.subplot(2, 2, 3)
        components = {
            'Repeatability': results['Percent Contribution']['Repeatability (EV)'],
            'Reproducibility': results['Percent Contribution']['Reproducibility (AV)'],
            'Part Variation': results['Percent Contribution']['Parts']
        }
        plt.bar(components.keys(), components.values())
        plt.title('Components of Variation')
        plt.ylabel('Percent Contribution')
        plt.grid()
    
        # Operator*Part Interaction
        plt.subplot(2, 2, 4)
        interaction_data = data.groupby([operator_col, part_col])[measurement_col].mean().unstack()
        for part in interaction_data.columns:
            plt.plot(interaction_data.index, interaction_data[part], 'o-', label=f'Part {part}')
        plt.title('Operator*Part Interaction')
        plt.xlabel('Operator')
        plt.ylabel('Mean Measurement')
        plt.legend()
        plt.grid()
        
        plt.tight_layout()
        plt.show()

    def get_sample_data():
        # Example data (typical Gage R&R study format)
        example_data= pd.DataFrame({
            'Operator': ['A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A',
                         'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B',
                         'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C'],
            'Part': [1, 1, 1, 2, 2, 2, 3, 3, 3, 4, 
                     1, 1, 1, 2, 2, 2, 3, 3, 3, 4,
                     1, 1, 1, 2, 2, 2, 3, 3, 3, 4],
            'Measurement': [21.2, 21.3, 21.4, 22.7, 22.8, 22.9, 23.0, 23.1, 23.2, 24.1,
                           21.1, 21.2, 21.3, 22.6, 22.7, 22.8, 23.1, 23.2, 23.3, 24.2,
                           21.3, 21.4, 21.5, 22.8, 22.9, 23.0, 23.2, 23.3, 23.4, 24.3]
        })
        
        return example_data    
```

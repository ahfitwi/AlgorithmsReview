### 1. Array
- ***Definition***: Contiguous memory locations storing elements of the same type (int only, float only, ...).
  - An array is a fundamental data structure that stores elements of the same data type in contiguous memory locations. Each element can be accessed directly using its index position.
- Use Cases: Storing sequential data, matrices, buffers, lookup tables.

- ***Key Characteristics***:
  - Fixed size (in most implementations)
  - Contiguous memory allocation
  - Zero-based indexing (in most programming languages)
  - Homogeneous elements (same data type)
  - Random access capability
- ***Use Cases***: Common Applications of Array Are:
  - Storing collections of data - lists of numbers, strings, or objects
  - Mathematical computations - matrices, vectors
  - Implementation of other data structures - stacks, queues, hash tables
  - Sorting and searching algorithms - binary search, various sorting algorithms
  - Buffer storage - temporary data storage
  - Image processing - pixel arrays
  - Database systems - record storage

- ***Advantages***:
  - Fast random access (O(1) time complexity)
  - Memory efficient (no overhead for pointers/links)
  - Cache friendly due to contiguous memory

- ***Disadvantages***:
  - Fixed size (in static arrays)
  - Insertion/deletion can be expensive (O(n))
  - Memory may be wasted if array is not fully utilized

- ***Python Implementation***:
  - Python doesn't have built-in arrays in the traditional sense, but provides several alternatives:

#### 3. Custom Array Implementation

```python
class CustomArray:
    def __init__(self, size):
        self.size = size
        self.data = [None] * size
    
    def __getitem__(self, index):
        if 0 <= index < self.size:
            return self.data[index]
        raise IndexError("Index out of range")
    
    def __setitem__(self, index, value):
        if 0 <= index < self.size:
            self.data[index] = value
        else:
            raise IndexError("Index out of range")
    
    def __len__(self):
        return self.size
    
    def __str__(self):
        return str(self.data)
    
    def search(self, value):
        """Linear search for value"""
        for i in range(self.size):
            if self.data[i] == value:
                return i
        return -1
    
    def reverse(self):
        """Reverse the array"""
        left, right = 0, self.size - 1
        while left < right:
            self.data[left], self.data[right] = self.data[right], self.data[left]
            left += 1
            right -= 1

# Usage example
arr = CustomArray(5)
arr[0] = 10
arr[1] = 20
arr[2] = 30
arr[3] = 40
arr[4] = 50

print("Custom array:", arr)
print("Element at index 2:", arr[2])
print("Search for 30:", arr.search(30))

arr.reverse()
print("After reverse:", arr)
```

### 2. Linked List
- Definition: Linear collection of nodes where each node points to the next.
  - **head**: The head of the linked list, which is the first node.
  - **tail**: The tail of the linked list, which is the last node.
  - **size**: The size of the linked list.
- The primary operations of a singly linked list include:
  - Traversal
  - Insertion
  - Deletion
- Use Cases: Dynamic memory allocation, stacks, queues, undo functionality.

#### What is a Linked List?
- A Linked List is a linear data structure where elements, called nodes, are stored in non-contiguous (not next to each other) memory locations. Unlike an array, where elements are stored in a single block of memory, linked list nodes are linked together using pointers (or references).
- Think of it like a scavenger hunt or a train: Each clue (node) tells you two things: 1) the item at that location, and 2) where to find the next clue.
- You have to follow the chain from the start to get to any specific point.

##### i. Basic Building Block: The Node
- A node is the fundamental unit of a linked list. It contains:
  - Data: The value you want to store (an integer, string, object, etc.).
  - Next (or Link): A pointer/reference to the next node in the sequence.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None  # Initially, it points to nothing
```

##### ii. 2. Types of Linked Lists
- There are several common types, each with its own use case:
- Singly Linked List: Each node points only to the next node. The last node points to None (or null). You can only traverse in one direction (from head to tail).
  - [Head] -> [A] -> [B] -> [C] -> None

- Doubly Linked List: Each node has two pointers: one to the next node and one to the previous node. This allows for traversal in both directions but uses more memory.
  - None <-> [A] <-> [B] <-> [C] <-> None

- Circular Linked List: The last node points back to the first node, forming a circle. It can be singly or doubly circular.
  - [Head] -> [A] -> [B] -> [C] -> [Head] (back to start)
 


##### iii. Key Operations and Their Complexities
- Let's focus on a Singly Linked List. We'll assume we have a LinkedList class that manages a head pointer.

|Operation	|Description	|Time Complexity	|Explanation|
|---|---|---|---|
|Access	|Find an element by its index.	|O(n)	|You must start at the head and traverse each node until you reach the desired index.|
|Search|	Find an element by its value.	|O(n)	|In the worst case, you may need to check every node.|
|Insertion|	- At the beginning	|O(1)	|You just create a new node and point it to the current head. Very fast!|
| | - At the end	|O(n)	|You have to traverse the entire list to find the last node before adding the new one.|
| |- At a specific position	|O(n)	|The traversal time to find the position dominates.|
|Deletion	|- From the beginning	|O(1)|	You just update the head to point to the second node.|
| |- From the end/middle	|O(n)	|You need to find the node before the one you want to delete, which requires traversal.|

- ***Singly LinkedList***

```python
#-------------------------------------------------------------------------------
# [Start]Linked List Noned Classs
#-------------------------------------------------------------------------------
class Node:
    #---------------------------------------------------------------------------
    # Dunder/Init Method
    #---------------------------------------------------------------------------
    def __init__(self, data):
        self.data = data
        self.next = None
        
#-------------------------------------------------------------------------------
# [End]Linked List Noned Classs
#-------------------------------------------------------------------------------

#-------------------------------------------------------------------------------
# [Start]Linked List Class
#-------------------------------------------------------------------------------
class LinkedList:
    #---------------------------------------------------------------------------
    # Dunder/Init Method
    #---------------------------------------------------------------------------
    def __init__(self):
        self.head = None

    #---------------------------------------------------------------------------
    # 1. Insert At The Beginning,  (TC=O(1), & SC=O(1))
    #---------------------------------------------------------------------------
    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node    

    #---------------------------------------------------------------------------
    # 2. Insert At The End (TC=O(n), & SC=O(1))
    #---------------------------------------------------------------------------
    def insert_at_end(self, data):
        new_node = Node(data)
        
        # If the list is empty, new node becomes the head
        if self.head is None:
            self.head = new_node
            return
        
        # Else, traverse to the last node
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node


    #---------------------------------------------------------------------------
    # 3. Insert after a specific node (TC=O(n), & SC=O(1))
    #---------------------------------------------------------------------------
    def insert_after_index(self, prev_data, data):
        if self.head is None:
            print("LinkedList is empty")
            return
        
        current = self.head
        while current:
            if current.data == prev_data:
                new_node = Node(data)
                new_node.next = current.next
                current.next = new_node
                return
            current = current.next
        
        print(f"Node with data {prev_data} not found")

    #---------------------------------------------------------------------------
    # 4. Delete a node by key (TC=O(n), & SC=O(1))
    #---------------------------------------------------------------------------
    def delete_node(self, key):
        temp = self.head

        # Case 1: Node to be deleted is the head
        if temp is not None and temp.data == key:
            self.head = temp.next
            temp = None
            return

        # Case 2: Find the node to be deleted
        prev = None
        while temp is not None and temp.data != key:
            prev = temp
            temp = temp.next

        # Case 3: Key was not found in the list
        if temp is None:
            print(f"Key {key} not found in the list.")
            return

        # Unlink the node from the list
        prev.next = temp.next
        temp = None

    #---------------------------------------------------------------------------
    #  5. Delete by value (TC=O(1), & SC=O(1))
    #---------------------------------------------------------------------------
    def delete_nodev(self, data):
        if self.head is None:
            print("LList is empty")
            return
        
        # If head node itself holds the key
        if self.head.data == data:
            self.head = self.head.next
            return
        
        current = self.head
        while current.next:
            if current.next.data == data:
                current.next = current.next.next
                return
            current = current.next
        
        print(f"Node with data {data} not found")


    #---------------------------------------------------------------------------
    # 6. Search operation (TC=O(n), & SC=O(1))
    #---------------------------------------------------------------------------
    def search(self, data):
        current = self.head
        position = 0
        
        while current:
            if current.data == data:
                return position
            current = current.next
            position += 1
        
        return -1


    #---------------------------------------------------------------------------
    # 7. Get Length of LL (TC=O(n), & SC=O(1)) 
    #---------------------------------------------------------------------------     
    def get_length(self):
        count = 0
        current = self.head
        
        while current:
            count += 1
            current = current.next
        
        return count


    #---------------------------------------------------------------------------
    # 8. Reverse the list (TC=O(n), & SC=O(1))
    #---------------------------------------------------------------------------        
    def reverse(self):
        prev = None
        current = self.head
        
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        
        self.head = prev


    #---------------------------------------------------------------------------
    # 9. Check if list is empty 
    #---------------------------------------------------------------------------        
    def is_empty(self):
        return self.head is None


    #---------------------------------------------------------------------------
    # 10. Get middle element 
    #---------------------------------------------------------------------------    
    def get_middle(self):
        if self.head is None:
            return None
        
        slow = self.head
        fast = self.head
        
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        
        return slow.data

    #---------------------------------------------------------------------------
    # 11.Detect cycle in LinkedList
    #---------------------------------------------------------------------------    
    def has_cycle(self):
        if not self.head:
            return False
        
        slow = self.head
        fast = self.head
        
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            
            if slow == fast:
                return True
        
        return False


    #---------------------------------------------------------------------------
    # 12.Find nth node from end 
    #---------------------------------------------------------------------------
    def find_nth_from_end(self, n):
        if not self.head:
            return None
        
        first = self.head
        second = self.head
        
        # Move first pointer n nodes ahead
        for _ in range(n):
            if not first:
                return None
            first = first.next
        
        # Move both pointers until first reaches end
        while first:
            first = first.next
            second = second.next
        
        return second.data if second else None
        
    #---------------------------------------------------------------------------
    # 13. Print the list (Traversal) (TC = O(n), SC = )    
    #---------------------------------------------------------------------------
    def __str__(self):
        nodes = []
        current = self.head
        if current is None:
            print("LinkedList is empty")
            return
            
        while current:
            nodes.append(str(current.data))
            current = current.next
        return " -> ".join(nodes)
        
    def display(self):
        if self.head is None:
            print("List is empty")
            return
        
        elements = []
        current = self.head
        while current:
            elements.append(str(current.data))
            current = current.next
        
        print(" -> ".join(elements))

#-------------------------------------------------------------------------------
# [End]Linked List Classs
#-------------------------------------------------------------------------------
```

- ***Circular Linked List***

```python
class CircularLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    
    # 1. Insert at beginning
    def insert_at_beginning(self, data):
        new_node = Node(data)
        
        if self.head is None:
            # Empty list - make it circular
            self.head = new_node
            self.tail = new_node
            new_node.next = self.head
        else:
            new_node.next = self.head
            self.head = new_node
            self.tail.next = self.head  # Maintain circularity
    
    # 2. Insert at end
    def insert_at_end(self, data):
        new_node = Node(data)
        
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            new_node.next = self.head
        else:
            self.tail.next = new_node
            new_node.next = self.head
            self.tail = new_node
    
    # 3. Insert after a specific node
    def insert_after(self, prev_data, data):
        if self.head is None:
            print("List is empty")
            return
        
        current = self.head
        while True:
            if current.data == prev_data:
                new_node = Node(data)
                new_node.next = current.next
                current.next = new_node
                
                # If inserting after tail, update tail
                if current == self.tail:
                    self.tail = new_node
                return
            
            current = current.next
            if current == self.head:  # Completed full circle
                break
        
        print(f"Node with data {prev_data} not found")
    
    # 4. Delete by value
    def delete_node(self, data):
        if self.head is None:
            print("List is empty")
            return
        
        # Case 1: Single node list
        if self.head == self.tail and self.head.data == data:
            self.head = None
            self.tail = None
            return
        
        # Case 2: Delete head node
        if self.head.data == data:
            self.head = self.head.next
            self.tail.next = self.head
            return
        
        # Case 3: Delete middle or tail node
        current = self.head
        prev = None
        
        while True:
            if current.data == data:
                # If deleting tail node, update tail
                if current == self.tail:
                    self.tail = prev
                
                prev.next = current.next
                return
            
            prev = current
            current = current.next
            
            if current == self.head:  # Completed full circle
                break
        
        print(f"Node with data {data} not found")
    
    # 5. Search operation
    def search(self, data):
        if self.head is None:
            return -1
        
        current = self.head
        position = 0
        
        while True:
            if current.data == data:
                return position
            current = current.next
            position += 1
            
            if current == self.head:  # Completed full circle
                break
        
        return -1
    
    # 6. Get length
    def get_length(self):
        if self.head is None:
            return 0
        
        count = 0
        current = self.head
        
        while True:
            count += 1
            current = current.next
            if current == self.head:
                break
        
        return count
    
    # 7. Check if list is empty
    def is_empty(self):
        return self.head is None
    
    # 8. Display the list
    def display(self):
        if self.head is None:
            print("List is empty")
            return
        
        elements = []
        current = self.head
        
        while True:
            elements.append(str(current.data))
            current = current.next
            if current == self.head:
                break
        
        print(" -> ".join(elements) + " → (head)")
    
    # 9. Traverse the list (circular traversal)
    def traverse(self, steps):
        if self.head is None:
            print("List is empty")
            return
        
        current = self.head
        result = []
        
        for _ in range(steps):
            result.append(str(current.data))
            current = current.next
        
        print(" → ".join(result) + " → ...")
    
    # 10. Check if list has cycle (always True for circular list)
    def has_cycle(self):
        return self.head is not None
    
    # 11. Split list into two halves
    def split_list(self):
        if self.head is None or self.head == self.tail:
            return None, None
        
        # Find middle using slow and fast pointers
        slow = self.head
        fast = self.head
        
        while fast.next != self.head and fast.next.next != self.head:
            slow = slow.next
            fast = fast.next.next
        
        # Create first half
        first_head = self.head
        first_tail = slow
        
        # Create second half
        second_head = slow.next
        second_tail = self.tail
        
        # Make both halves circular
        first_tail.next = first_head
        second_tail.next = second_head
        
        # Create new lists
        first_list = CircularLinkedList()
        first_list.head = first_head
        first_list.tail = first_tail
        
        second_list = CircularLinkedList()
        second_list.head = second_head
        second_list.tail = second_tail
        
        return first_list, second_list
    
    # 12. Josephus Problem solution
    def josephus_problem(self, k):
        """
        Josephus problem: Eliminate every kth person until one remains
        Returns the last remaining node's data
        """
        if self.head is None:
            return None
        
        if k <= 0:
            return None
        
        current = self.head
        
        # Continue until only one node remains
        while self.head != self.tail:
            # Move k-1 steps
            for _ in range(k - 1):
                current = current.next
            
            # The next node is the one to eliminate
            node_to_delete = current.next
            
            # If we're deleting head, update head
            if node_to_delete == self.head:
                self.head = self.head.next
            
            # If we're deleting tail, update tail
            if node_to_delete == self.tail:
                self.tail = current
            
            # Bypass the node to delete
            current.next = node_to_delete.next
            
            # Move to next node for next iteration
            current = current.next
        
        return self.head.data
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_node = Node(value)
        cur_node = self.head

        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return

        while cur_node.next != self.head:
            cur_node = cur_node.next
        cur_node.next = new_node
        new_node.next = self.head

    def prepend(self, value):
        cur = self.head
        new_node = Node(value)
        new_node.next = self.head

        if not self.head:
            new_node.next = new_node
        else:
            while cur.next != self.head:
                cur = cur.next
            cur.next = new_node
        self.head = new_node

    def remove(self, key):
        if self.head:
            if self.head.value == key:
                cur = self.head
                while cur.next != self.head:
                    cur = cur.next
                if self.head == self.head.next:
                    self.head = None
                else:
                    cur.next = self.head.next
                    self.head = self.head.next
            else:
                cur = self.head
                prev = None
                while cur.next != self.head:
                    prev = cur
                    cur = cur.next
                    if cur.value == key:
                        prev.next = cur.next
                        cur = cur.next

    def __str__(self):
        cur = self.head
        linkedlist = ""

        while cur:
            linkedlist += str(cur.value) + "->"
            cur = cur.next

            if cur == self.head:
                return linkedlist

        return linkedlist


c = CircularLinkedList()
c.append(10)
c.append(20)
c.prepend(5)
c.remove(5)

print(c)
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.previous = None
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def __len__(self) -> int:
        return self.size

    def __str__(self) -> str:
        linkedlist = ""
        current  = self.head

        while current and current.next is not None:
            linkedlist += str(current.value) + "<->"
            current = current.next
        linkedlist += str(current.value)
        return linkedlist

    def prepend(self, element) -> None:
        new_node = Node(element)
        new_node.next = self.head

        if self.head is not None:
            self.head.previous = new_node

        self.head = new_node

        if self.tail == None:
            self.tail = self.head

        self.size += 1

    def append(self, element) -> None:
        if self.tail is None:
            self.prepend(element)
            return


        new_node = Node(element)

        self.tail.next = new_node

        new_node.previous = self.tail

        self.tail = new_node

        self.size += 1
    
    def add_at_position(self, position, element):
        if position < 1 or position > self.size:
            raise IndexError("Position is invalid")

        if position == 1:
            self.prepend(element)
            return

        index = 1
        current = self.head
        new_node = Node(element)

        while index != position:
            index += 1
            current = current.next

        prev = current.previous

        prev.next = new_node
        new_node.previous = prev
        new_node.next = current

        current.previous = new_node

        self.size += 1
    
    def delete_node(self, position):
        if position < 1 or position > self.size:
            raise IndexError("Invalid position!")

       
        if position == 1:
            node_to_remove = self.head

            new_head = self.head.next

            node_to_remove.next = None
            node_to_remove.previous = None

            self.head = new_head

            self.size -= 1

            return node_to_remove.value

        # Check if element to be removed is the tail
        if position == self.size:
            node_to_remove = self.tail
            new_tail = self.tail.previous

            node_to_remove.previous = None
            new_tail.next = None

            self.tail = new_tail

            self.size -= 1

            return node_to_remove.value

        index = 1
        current = self.head

        while index != position:
            index += 1
            current = current.next

        prev = current.previous

        prev.next = current.next

        current.next.previous = prev

        self.size -= 1

        return current.value

dll = LinkedList()
dll.prepend(40)
dll.prepend(30)
dll.append(50)
dll.add_at_position(1, 10)
dll.delete_node(2)

print(dll, len(dll))
```

#### Complete Stack Class

```python
class Stack:
    def __init__(self):
        self.items = []
    
    def push(self, item):
        self.items.append(item)
    
    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        raise IndexError("Pop from empty stack")
    
    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        raise IndexError("Peek from empty stack")
    
    def is_empty(self):
        return len(self.items) == 0
    
    def size(self):
        return len(self.items)
    
    def __str__(self):
        return str(self.items)

# Usage
stack = Stack()
stack.push(10)
stack.push(20)
stack.push(30)
print(stack)  # [10, 20, 30]

print(stack.pop())   # 30
print(stack.peek())  # 20
print(stack.size())  # 2
print(stack) # [10, 20]
```

#### Common Stack Operations
##### 1. Reverse a string

```python
def reverse_string(s):
    stack = []
    # Push all characters
    for char in s:
        stack.append(char)
    
    # Pop all characters (they come out in reverse order)
    reversed_str = ''
    while stack:
        reversed_str += stack.pop()
    
    return reversed_str

print(reverse_string("hello"))  # "olleh"
```

##### 2. Balanced parentheses

```python
def is_balanced(expression):
    stack = []
    matching = {')': '(', '}': '{', ']': '['}
    
    for char in expression:
        if char in '({[':
            stack.append(char)
        elif char in ')}]':
            if not stack or stack.pop() != matching[char]:
                return False
    
    return len(stack) == 0

print(is_balanced("({[]})"))  # True
print(is_balanced("({[})"))   # False
```

##### 3. Postfix expression evaluation

```python
def evaluate_postfix(expression):
    stack = []
    operators = set(['+', '-', '*', '/'])
    
    for token in expression.split():
        if token not in operators:
            stack.append(float(token))
        else:
            b = stack.pop()
            a = stack.pop()
            if token == '+':
                stack.append(a + b)
            elif token == '-':
                stack.append(a - b)
            elif token == '*':
                stack.append(a * b)
            elif token == '/':
                stack.append(a / b)
    
    return stack[0]

print(evaluate_postfix("3 4 + 2 *"))  # 14.0
```

##### Advanced Stack Implementation with Size Limit

```python
class FixedSizeStack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.items = []
    
    def push(self, item):
        if len(self.items) >= self.capacity:
            raise Exception("Stack overflow")
        self.items.append(item)
    
    def pop(self):
        if self.is_empty():
            raise Exception("Stack underflow")
        return self.items.pop()
    
    def peek(self):
        if self.is_empty():
            raise Exception("Stack is empty")
        return self.items[-1]
    
    def is_empty(self):
        return len(self.items) == 0
    
    def is_full(self):
        return len(self.items) == self.capacity
    
    def size(self):
        return len(self.items)

# Usage
fixed_stack = FixedSizeStack(3)
fixed_stack.push(1)
fixed_stack.push(2)
fixed_stack.push(3)
print(fixed_stack.is_full())  # True
```

```python
# Custom implementation
class HashTable:
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]
    
    def _hash(self, key):
        return hash(key) % self.size
    
    def put(self, key, value):
        index = self._hash(key)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value)
                return
        self.table[index].append((key, value))
    
    def get(self, key):
        index = self._hash(key)
        for k, v in self.table[index]:
            if k == key:
                return v
        return None

ht = HashTable()
ht
```

### 7. Binary Search Tree (BST)
- Definition: Binary tree where left child < parent < right child.
- Use Cases: Searching, sorting, database indexing.

```python
class BSTNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None
    
    def insert(self, value):
        if not self.root:
            self.root = BSTNode(value)
        else:
            self._insert_recursive(self.root, value)
    
    def _insert_recursive(self, node, value):
        if value < node.value:
            if node.left is None:
                node.left = BSTNode(value)
            else:
                self._insert_recursive(node.left, value)
        else:
            if node.right is None:
                node.right = BSTNode(value)
            else:
                self._insert_recursive(node.right, value)
    
    def search(self, value):
        return self._search_recursive(self.root, value)
    
    def _search_recursive(self, node, value):
        if node is None or node.value == value:
            return node
        if value < node.value:
            return self._search_recursive(node.left, value)
        return self._search_recursive(node.right, value)
```

```python
class GraphMatrix:
    def __init__(self, num_vertices, directed=False):
        self.num_vertices = num_vertices
        self.directed = directed
        self.matrix = [[0] * num_vertices for _ in range(num_vertices)]
    
    def add_edge(self, v1, v2, weight=1):
        self.matrix[v1][v2] = weight
        if not self.directed:
            self.matrix[v2][v1] = weight
    
    def remove_edge(self, v1, v2):
        self.matrix[v1][v2] = 0
        if not self.directed:
            self.matrix[v2][v1] = 0
    
    def get_neighbors(self, vertex):
        neighbors = []
        for i in range(self.num_vertices):
            if self.matrix[vertex][i] != 0:
                neighbors.append((i, self.matrix[vertex][i]))
        return neighbors
    
    def __str__(self):
        return '\n'.join([' '.join(map(str, row)) for row in self.matrix])

# Example usage
g = GraphMatrix(5)
g.add_edge(0, 1)
g.add_edge(0, 2)
g.add_edge(1, 3)
print("Adjacency Matrix:")
print(g)
```

### 2. Adjacency List

```python
from collections import defaultdict, deque

class Graph:
    def __init__(self, directed=False):
        self.graph = defaultdict(dict)
        self.directed = directed
        self.vertices = set()
    
    def add_vertex(self, vertex):
        self.vertices.add(vertex)
        if vertex not in self.graph:
            self.graph[vertex] = {}
    
    def add_edge(self, v1, v2, weight=1):
        self.add_vertex(v1)
        self.add_vertex(v2)
        
        self.graph[v1][v2] = weight
        if not self.directed:
            self.graph[v2][v1] = weight
    
    def remove_edge(self, v1, v2):
        if v1 in self.graph and v2 in self.graph[v1]:
            del self.graph[v1][v2]
            if not self.directed and v2 in self.graph and v1 in self.graph[v2]:
                del self.graph[v2][v1]
    
    def get_neighbors(self, vertex):
        return list(self.graph[vertex].items()) if vertex in self.graph else []
    
    def get_vertices(self):
        return list(self.vertices)
    
    def bfs(self, start_vertex):
        visited = set()
        queue = deque([start_vertex])
        result = []
        
        while queue:
            vertex = queue.popleft()
            if vertex not in visited:
                visited.add(vertex)
                result.append(vertex)
                for neighbor, _ in self.get_neighbors(vertex):
                    if neighbor not in visited:
                        queue.append(neighbor)
        return result
    
    def dfs(self, start_vertex):
        visited = set()
        result = []
        
        def dfs_recursive(vertex):
            visited.add(vertex)
            result.append(vertex)
            for neighbor, _ in self.get_neighbors(vertex):
                if neighbor not in visited:
                    dfs_recursive(neighbor)
        
        dfs_recursive(start_vertex)
        return result
    
    def dijkstra(self, start_vertex):
        import heapq
        
        distances = {vertex: float('inf') for vertex in self.vertices}
        distances[start_vertex] = 0
        priority_queue = [(0, start_vertex)]
        
        while priority_queue:
            current_distance, current_vertex = heapq.heappop(priority_queue)
            
            if current_distance > distances[current_vertex]:
                continue
            
            for neighbor, weight in self.get_neighbors(current_vertex):
                distance = current_distance + weight
                
                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    heapq.heappush(priority_queue, (distance, neighbor))
        
        return distances
    
    def has_cycle(self):
        visited = set()
        recursion_stack = set()
        
        def dfs_cycle(vertex):
            visited.add(vertex)
            recursion_stack.add(vertex)
            
            for neighbor, _ in self.get_neighbors(vertex):
                if neighbor not in visited:
                    if dfs_cycle(neighbor):
                        return True
                elif neighbor in recursion_stack:
                    return True
            
            recursion_stack.remove(vertex)
            return False
        
        for vertex in self.vertices:
            if vertex not in visited:
                if dfs_cycle(vertex):
                    return True
        return False
    
    def __str__(self):
        return '\n'.join([f"{v}: {list(neighbors.items())}" 
                         for v, neighbors in self.graph.items()])

# Example usage
g = Graph()
g.add_edge('A', 'B', 4)
g.add_edge('A', 'C', 2)
g.add_edge('B', 'C', 1)
g.add_edge('B', 'D', 5)
g.add_edge('C', 'D', 8)

print("Graph (Adjacency List):")
print(g)

print("\nBFS from A:", g.bfs('A'))
print("DFS from A:", g.dfs('A'))
print("Dijkstra from A:", g.dijkstra('A'))
print("Has cycle:", g.has_cycle())
```

### 3. Weighted Graph with Path Finding

```python
class WeightedGraph:
    def __init__(self, directed=False):
        self.graph = defaultdict(dict)
        self.directed = directed
    
    def add_edge(self, u, v, weight):
        self.graph[u][v] = weight
        if not self.directed:
            self.graph[v][u] = weight
    
    def shortest_path(self, start, end):
        import heapq
        
        distances = {node: float('inf') for node in self.graph}
        previous = {node: None for node in self.graph}
        distances[start] = 0
        heap = [(0, start)]
        
        while heap:
            current_distance, current_node = heapq.heappop(heap)
            
            if current_node == end:
                path = []
                while current_node is not None:
                    path.append(current_node)
                    current_node = previous[current_node]
                return path[::-1], current_distance
            
            if current_distance > distances[current_node]:
                continue
            
            for neighbor, weight in self.graph[current_node].items():
                distance = current_distance + weight
                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    previous[neighbor] = current_node
                    heapq.heappush(heap, (distance, neighbor))
        
        return None, float('inf')

# Example usage
wg = WeightedGraph()
wg.add_edge('A', 'B', 4)
wg.add_edge('A', 'C', 2)
wg.add_edge('B', 'C', 1)
wg.add_edge('B', 'D', 5)
wg.add_edge('C', 'D', 8)
wg.add_edge('C', 'E', 10)
wg.add_edge('D', 'E', 2)

path, distance = wg.shortest_path('A', 'E')
print(f"Shortest path from A to E: {path} with distance {distance}")
```

### 10. Trie
- Definition: Tree-like structure for efficient string searching.
- Use Cases: Autocomplete, spell check, IP routing.

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False

class Trie:
    def __init__(self):
        self.root = TrieNode()
    
    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end = True
    
    def search(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        return node.is_end
    
    def starts_with(self, prefix):
        node = self.root
        for char in prefix:
            if char not in node.children:
                return False
            node = node.children[char]
        return True
```

## <div class="alert alert-info"> 1. Prefix Sum

- ***Definition***: Each element $p_{i}$ in the prefix sum sequence is calculated as the sum of elements from $a_{0}$ to $a_{i}$ in the original sequence:
  $$p_i\ =\ \sum_{k=1}^i a_k$$
- A prefix sum (or cumulative sum or scan sum) is a running total of a sequence of numbers, where each element in the prefix sum array is the sum of all preceding elements in the original array up to that point.
- This technique allows for efficient calculation of the sum of any subarray in constant time (\(O(1)\)) after an initial \(O(n)\) preprocessing step to build the prefix sum array.
- ***Monotonicity (Conditional)***: If all elements in the original sequence $A$ are non-negative, the prefix sum sequence $P$ will be non-decreasing. If all elements in $A$ are positive, $P$ will be strictly increasing.
- ***Zero Subarray***: If the sequence $p_i$ contains duplicates, there exists a zero subarray. In essence, the presence of zero-sum subarrays manifests as duplicate values within the prefix sum sequence. The indices of these duplicate values directly correspond to the start and end points of the zero-sum subarrays.
- ***Inverse Relationship with Finite Differences***: The original sequence can be recovered from its prefix sum sequence by taking finite differences. Specifically, $$a_{i}=p_{i}-p_{i-1}\  for\ i>0,\  and\ a_{0}=p_{0}$$  


```python
def calc_prefix_sum(input_arr=None, inplace=False):
    n=len(input_arr)
    if inplace:
        for k in range(1, n):
            input_arr[k] = input_arr[k-1]+input_arr[k]
        
        return input_arr
    else:
        ps = [0]*n
        ps[0] = input_arr[0]
        for k in range(1, n):
            ps[k] = ps[k-1]+input_arr[k]
            
        return ps

arr = [1, 2, 3, 4, 5, 6, 7]
ps = calc_prefix_sum(input_arr=arr)

print(f"arr = {arr}\n")
print(f"ps = {ps}\n")
```

```python
arr = [1, 2, 3, 4, 5, 6, 7]
ps = calc_prefix_sum(input_arr=arr, inplace=True)

print(f"arr = {arr}\n")
print(f"ps = {ps}\n")
```

### 1. Range Sum Queries (Most Common Use Case)

```python
# O(n*k)
def find_subarray_sum_n(arr, i, j):
    sub_arr_sum=0;

    for k in range(i, j+1):
        sub_arr_sum += arr[k]

    return sub_arr_sum

arr = [1, 2, 3, 4, 5, 6, 7]
find_subarray_sum_n(arr, 3, 6)
```

```python
def calc_range_sum_qquery(arr, i, j):
    # Precompute prefix sum
    ps = calc_prefix_sum(input_arr=arr, inplace=False)

    if i>0:
        return ps[j]-ps[i-1]
    else:
        if j<len(arr):
            return ps[j]
        else:
            return -1

arr = [3, 7, 2, 5, 8]
calc_range_sum_qquery(arr, 0, 5)
```

```python
def subarray_with_sum(arr, target):
    prefix_sum = 0
    prefix_map = {0: -1}  # sum 0 at index -1
    
    for i, num in enumerate(arr):
        prefix_sum += num
        
        # Check if we've seen (prefix_sum - target) before
        if prefix_sum - target in prefix_map:
            start = prefix_map[prefix_sum - target] + 1
            return arr[start:i + 1]
        
        prefix_map[prefix_sum] = i
    
    return None

arr = [1, 4, 20, 3, 10, 5]
target = 33
result = subarray_with_sum(arr, target)
print(f"Subarray with sum {target}: {result}")  # [20, 3, 10]
```

### 3. Equilibrium Point Finding
- Find index where sum of left elements equals sum of right elements.

```python
def find_equilibrium(arr):
    total_sum = sum(arr)
    left_sum = 0
    
    for i, num in enumerate(arr):
        # right_sum = total_sum - left_sum - num
        if left_sum == total_sum - left_sum - num:
            return i
        left_sum += num
    
    return -1

arr = [-7, 1, 5, 2, -4, 3, 0]
print(f"Equilibrium index: {find_equilibrium(arr)}")  # 3
```

### 4. Split an array into two equal Sum subarrays
- Given an array of integers greater than zero, find if it is possible to split it in two subarrays (without reordering the elements), such that the sum of the two subarrays is the same. Print the two subarrays.
- Equilibrium Index

```python
def split_arr_to_equal_parts(arr, use_ps=0):
    n=len(arr)
    if use_ps==0:
        tot_sum = sum(arr);
        for i in range(n):
            if 2*sum(arr[:i]) == tot_sum :
                return arr[:i], arr[i:]
        
    elif use_ps==1:
        # Precompute Prefix Sum
        ps = calc_prefix_sum(input_arr=arr, inplace=False)
        for i in range(n):
            if 2*ps[i] == ps[-1]:
                return arr[:i], arr[i:]
        
    else:
        for i in range(n):
            if sum(arr[:i]) == sum(arr[i:]):
                return arr[:i], arr[i:]
            
    return -1   
```

### 5. Mean of range in array
- Given an array arr[] of n integers and q queries represented by an array queries[][]
- , where queries[i][0] = l and queries[i][1] = r. For each query, the task is to calculate the mean of elements in the range l to r and return its floor value.
    - Input: arr = [3, 7, 2, 8, 5] queries = [[1, 3], [2, 5]]
    - Output: 4 5

```python
def calc_mean_range(arr, queries):
    # Precompute Prefix Sum In O(n)
    ps = calc_prefix_sum(input_arr=arr, inplace=False)

    # Means
    means, n = [], len(arr)
    for l, r in queries:
        if l>r:
            means.append(-1)
        else:
            if r<0:
                r = r if r>=-1*n else -1*n
            else:
                r = r if r<n else n-1

            if l<0:
                l = l if l>=-1*n else -1*n
            else:
                l = l if l<n else n-1     
        if l==0:
            tmp = ps[r]
        else:
            tmp = ps[r]-ps[l-1]
            
        if tmp == -1:
            means.append(-1)
        else:
            means.append(int(tmp/(r-l+1)))


    return means    
```

### 6. Original Array from given Prefix Sums
- You are given a prefix sum array presum[] of an array arr[]. The task is to find the original array arr[] whose prefix sum is presum[].
  - Input: {5, 7, 10, 11, 18}
  - Output: [5, 2, 3, 1, 7]

```python
def recover_arr_from_prefix_sum(ps, inplace=False):
    n=len(ps)     
    if inplace:
        tmp = ps[0]
        for k in range(n-1, 0, -1):
            ps[k] = ps[k]-ps[k-1]
        ps[0] = tmp
        
        return ps
    else:
        arr = [ps[0]]
        for k in range(1, n):
            arr.append(ps[k]-ps[k-1])
            
        return arr

ps = [5, 7, 10, 11, 18]
recover_arr_from_prefix_sum(ps, inplace=False)
```

### 7. Maximum Subarray Sum (Kadane's Algorithm)

```python
def max_subarray_sum(arr):
    max_ending_here = arr[0]
    max_so_far = arr[0]
    
    for num in arr[1:]:
        max_ending_here = max(num, max_ending_here + num)
        max_so_far = max(max_so_far, max_ending_here)
    
    return max_so_far

arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
arr = [20, -2, 1, -3, 4, -1, 2, 1, -5, 4]
print(f"Maximum subarray sum: {max_subarray_sum(arr)}")  # 6
```

### 9. Longest Subarray With Sum K
- Given an array arr[] of size n containing integers, the task is to find the length of the longest subarray having sum equal to the given value k.
- Note: If there is no subarray with sum equal to k, return 0.
- Examples: 

        Input: arr[] = [10, 5, 2, 7, 1, -10], k = 15
        Output: 6
        Explanation: Subarrays with sum = 15 are [5, 2, 7, 1], [10, 5] and [10, 5, 2, 7, 1, -10]. The length of the longest subarray with a sum of 15 is 6.
        
        Input: arr[] = [-5, 8, -14, 2, 4, 12], k = -5
        Output: 5
        Explanation: Only subarray with sum = 15 is [-5, 8, -14, 2, 4] of length 5.
        
        Input: arr[] = [10, -10, 20, 30], k = 5
        Output: 0
        Explanation: No subarray with sum = 5 is present in arr[].

- **Solution**:
  - Nested Loop - O(n^2) Time and O(1) Space
  - Using Hash Map and Prefix Sum - O(n) Time and O(n) Space

```python
def longestSubarray(arr, k):
    res = 0

    for i in range(len(arr)):
        
        # Sum of subarray from i to j
        sum1 = 0
        for j in range(i, len(arr)):
            sum1 += arr[j]
          
            # If subarray sum is equal to k
            if sum1 == k:
              
                # find subarray length and update result
                subLen = j - i + 1
                res = max(res, subLen)
    
    return res

arr = [10, 5, 2, 7, 1, -10]
k = 15
longestSubarray(arr, k)
```

```python
def find_longest_subarray(arr, k):
    # Init vars
    hmp, cnt, prefix_sum, n = {}, 0, 0, len(arr)
    
    for i in range(n):
        prefix_sum += arr[i]

        # Check if the entire prefix sums to k
        if prefix_sum == k:
            cnt = i + 1

        # If prefixSum - k exists in the map then there exist such 
      	# subarray from (index of previous prefix + 1) to i.
        elif (prefix_sum - k) in hmp:
            cnt = max(cnt, i - hmp[prefix_sum - k])

        # Store only first occurrence index of prefSum
        if prefix_sum not in hmp:
            hmp[prefix_sum] = i

    return cnt

arr = [10, 5, 2, 7, 1, -10]
k = 15
print(find_longest_subarray(arr, k))
```

```python
def find_longest_subarray2(arr, k):
    mp, res, n = {}, 0, len(arr)
    
    # Precompute Prefix Sum In O(n)
    ps = calc_prefix_sum(input_arr=arr, inplace=False)

    for i in range(n):

        # Check if ith prefix sums to k
        if ps[i] == k:
            res = i + 1

        # If ps[i] - k exists in the map then there exist such 
      	# subarray from (index of previous prefix + 1) to i.
        elif (ps[i] - k) in mp:
            res = max(res, i - mp[ps[i]- k])

        # Store only first occurrence index of prefSum
        if ps[i] not in mp:
            mp[ps[i]] = i

    return res

arr = [10, 5, 2, 7, 1, -10]
k = 15
print(find_longest_subarray2(arr, k))
```

### 10. Minimum Removals for Target Sum
- Given an array of positive integers arr[] and an integer k, you can remove either the leftmost or rightmost element from the array in one operation. After each operation, the size of arr[] is reduced by one. find the minimum number of operations required to make the total sum of the removed elements exactly equal to k. If it is impossible to achieve the sum k, return -1.

- Examples : 

        Input: arr[] = [3, 4, 1, 3, 2], k = 5
        Output: 2
        Explanation: Removing 3 from left and 2 from right gives a sum of 5 in 2 operations.
        
        Input: arr[] = [5, 3, 4, 6, 2], k = 6
        Output: -1
        Explanation: It is impossible to achieve the sum of removed elements as 6.
        
        Input: arr[] = [1, 1, 3, 1, 2], k = 4
        Output: 3
        Explanation: Removing 1, 1 from left and 2 from right gives a sum of 4 in 3 operations.

- [Naive Approach]: Recursive Approach - O(2 ^ n) Time and O(n) Space
  - The idea is to use recursion to explore all combinations of removing elements from the left or right. In each recursive call, we have two choices: either to remove the left most element or remove the right most element. After exploring both the choices, return the minimum of the two. If all elements are exhausted, return -1.

```python
def minRemovalsRec(arr, k, left, right, cnt):

    # Target sum achieved
    if k == 0:
        return cnt

    # No elements left
    if left > right:
        return float('inf')

    # Remove leftmost element
    l = minRemovalsRec(arr, k - arr[left], left + 1, right, cnt + 1)

    # Remove rightmost element
    r = minRemovalsRec(arr, k - arr[right], left, right - 1, cnt + 1)

    return min(l, r)

def minRemovals(arr, k):
    res = minRemovalsRec(arr, k, 0, len(arr) - 1, 0)
    return -1 if res == float('inf') else res


arr = [3, 4, 1, 3, 2]
k = 5
print(minRemovals(arr, k))
```

- [Better Approach] Using Prefix Array - O(n) Time and O(n) Space
  - If we observe carefully, we can say that after removing the elements whose sum = k, the sum of the remaining elements will be (total sum - k). Since we need to minimize the removals, the problem can be reduced to finding the longest subarray whose sum = (total sum - k).
  - We can find the longest subarray having sum = (total sum - k) by using a hash map or dictionary to store the first occurrence of each prefix sum. For each index i, we find the prefix sum till index i, say prefSum and find the first occurrence of (prefSum - (total sum - k)) in the hash map, if present. The difference between the first occurrence and the current index i will be the length of the subarray. 

```python
def calc_min_removals(arr, k):
    # Init vars
    hmp, res, prefix_sum, n = {}, 0, 0, len(arr)

    # Update Target Sum
    k = sum(arr)-k;
    
    for i in range(n):
        prefix_sum += arr[i]

        # Check if the entire prefix sums to k
        if prefix_sum == k:
            res = i + 1

        # If prefixSum - k exists in the map then there exist such 
      	# subarray from (index of previous prefix + 1) to i.
        elif (prefix_sum - k) in hmp:
            res = max(res, i - hmp[prefix_sum - k])

        # Store only first occurrence index of prefSum
        if prefix_sum not in hmp:
            hmp[prefix_sum] = i

    return n-res

arr = [3, 4, 1, 3, 2]
k = 5
calc_min_removals(arr, k)
```

### 11. Subarray with 0 sum
- Given an array of positive and negative numbers, the task is to find if there is a subarray (of size at least one) with 0 sum.
- Examples: 

            Input: {4, 2, -3, 1, 6}
            Output: true 
            Explanation:
            There is a subarray with zero sum from index 1 to 3.
            
            Input: {4, 2, 0, 1, 6}
            Output: true
            Explanation: The third element is zero. A single element is also a sub-array.
            
            Input: {-3, 2, 3, 1, 6}
            Output: false

- [Naive Method]: Subarray with 0 sum using Nested loop: Generate every subarray and calcuate the sum of each subarray. Check if subarray sum is 0 then return true. Otherwise, if no such subarray found then return false.
  - Time Complexity: $O(n^2)$
  - Auxiliary Space: $O(1)$


```python
def isSubArrayExists(arr, target=0):

    # Length of Array
    n = len(arr)

    # Check if Target Sum Exists
    for i in range(n):
        # Starting point of the subarray and
        # sum is initialized with the first element of subarray
        sum1 = arr[i]
        if sum1 == 0:
            return True
        for j in range(i + 1, n):
            # We are finding the sum till the jth index
            # starting from the ith index
            sum1 += arr[j]
            if sum1 == target:
                return True
    return False

# Driver's code
if __name__ == "__main__":
    arr = [-3, 2, 3, 1, 6]
    
    # Function call
    if isSubArrayExists(arr, target=0):
        print("Found a subarray with 0 sum")
    else:
        print("No Such Sub Array Exists!")
```

### Using Hashing:
- The idea is to iterate through the array and for every element arr[i], calculate the sum of elements from 0 to i (this can simply be done as sum += arr[i]). If the current sum has been seen before, then there must be a zero-sum subarray. Hashing is used to store the sum values so that sum can be stored quickly and find out whether the current sum is seen before or not.
- $O(n)$
- Follow the given steps to solve the problem:
  - Declare a variable sum, to store the sum of prefix elements
  - Traverse the array and at each index, add the element into the sum and check if this sum exists earlier. If the sum exists, then return true
  - Also, insert every prefix sum into a map, so that later on it can be found whether the current sum is seen before or not
  - At the end return false, as no such subarray is found
- Illustration:

            arr[] = {1, 4, -2, -2, 5, -4, 3}
            
            Consider all prefix sums, one can notice that there is a subarray with 0 sum when :
            
            Either a prefix sum repeats
            Or, prefix sum becomes 0.
            Prefix sums for above array are: 1, 5, 3, 1, 6, 2, 5
            Since prefix sum 1 repeats, we have a subarray with 0 sum. 

```python
def find_subarray_zerosum(arr=None, k=0):
    # Init vars
    hmp, res, prefix_sum, n = {}, 0, 0, len(arr)
    
    for i in range(n):
        prefix_sum += arr[i]

        # Check if the entire prefix sums to k
        if prefix_sum == k:
            res = i + 1

        # If prefixSum - k exists in the map then there exist such 
      	# subarray from (index of previous prefix + 1) to i.
        elif (prefix_sum - k) in hmp:
            res = i+1

        # Store only first occurrence index of prefSum
        if prefix_sum not in hmp:
            hmp[prefix_sum] = i

        if res > 0:
            return True

    
    return False


print(find_subarray_zerosum(arr=[4, 2, -3, 1, 6], k=0))
print(find_subarray_zerosum(arr=[4, 2, 0, 1, 6], k=0))
print(find_subarray_zerosum(arr=[-3, 2, 3, 1, 6], k=0))
```

```python
# python3 program to find if
# there is a zero sum subarray
def isSubArrayExists(arr, target=0):
    # traverse through array
    # and store prefix sums
    n_sum, s, N = 0, set(), len(arr)
    s = set()

    for i in range(N):
        n_sum += arr[i]

        # If prefix sum is 0 or
        # it is already present
        if n_sum == target or n_sum in s:
            return True
        s.add(n_sum)

    return False


# Driver's code
if __name__ == '__main__':
    arr = [-3, 2, 3, 1, 6]
    
    # Function call
    if isSubArrayExists(arr, target=0) == True:
        print("Found a subarray with 0 sum")
    else:
        print("No Such sub array exits!")
```

### 12. Largest Submatrix With Sum 0
- Given a 2D matrix of dimension n x m, the task is to find the area of largest submatrix whose sum is 0.
- Examples: 

        Input: mat[][] = [[9,  7, 16,  5], [1, -6, -7,  3], [1,  8,  7,  9], [7, -2,  0, 10]]

```python
# Python program to find Largest rectangular
# sub-matrix whose sum is 0

def maxZeroSumSubarray(arr):
    prefSum = 0
    maxLength = 0

    # Hash map to store the first index of each prefix sum
    mp = {}

    # Iterate through the array to find subarrays with zero sum
    for i in range(len(arr)):
        prefSum += arr[i]

        if prefSum == 0:
            maxLength = i + 1

        if prefSum in mp:
            # If this prefSum repeats, find subarray length.
            maxLength = max(maxLength, i - mp[prefSum])
        else:
            # Only store the index of the first occurrence of prefSum
            mp[prefSum] = i

    return maxLength

def zeroSumSubmat(mat):
    rows = len(mat)
    cols = len(mat[0])
    maxArea = 0

    for i in range(rows):

        # Temporary array to store the column-wise cumulative sum
        temp = [0] * cols

        # Iterate over each row from i to j
        for j in range(i, rows):

            # Accumulate the column-wise sum for rows between i and j
            for k in range(cols):
                temp[k] += mat[j][k]

            # Find the longest zero-sum subarray in column sums
            lenSubarray = maxZeroSumSubarray(temp)

            # Update the maximum area 
            maxArea = max(maxArea, (j - i + 1) * lenSubarray)

    return maxArea

if __name__ == '__main__':
    mat = [
        [9, 7, 16, 5], 
        [1, -6, -7, 3], 
        [1, 8, 7, 9], 
        [7, -2, 0, 10]
    ]
    print(zeroSumSubmat(mat))
```

### 13. Matrix Range Sum Queries

```python
def matrix_range_sum(matrix, row1, col1, row2, col2):
    # Precompute 2D prefix sum
    m, n = len(matrix), len(matrix[0])
    prefix = [[0] * (n + 1) for _ in range(m + 1)]
    
    for i in range(m):
        for j in range(n):
            prefix[i + 1][j + 1] = (matrix[i][j] + 
                                   prefix[i][j + 1] + 
                                   prefix[i + 1][j] - 
                                   prefix[i][j])
    
    # Calculate range sum
    return (prefix[row2 + 1][col2 + 1] - 
            prefix[row1][col2 + 1] - 
            prefix[row2 + 1][col1] + 
            prefix[row1][col1])

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
# Sum from (1,1) to (2,2) → 5+6+8+9 = 28
print(f"Matrix range sum: {matrix_range_sum(matrix, 1, 1, 2, 2)}")
```

```python
def find_max_non_overlapping_areas(
                                    matrix: np.ndarray, 
                                    area_size: int = 40, 
                                    num_areas: int = 5
                                  ) -> List[Tuple[int, int]]:
    """
    Find the top non-overlapping areas with maximum sums in a 2D matrix.
    
    Args:
        matrix: 2D numpy array
        area_size: size of each square area (default 40)
        num_areas: number of areas to find (default 5)
    
    Returns:
        List of (row, col) indices representing top-left corners of areas
    """
    m, n = matrix.shape
    
    # Step 1: Precompute 2D prefix sum for efficient range sum queries
    prefix = np.zeros((m + 1, n + 1))
    for i in range(m):
        for j in range(n):
            prefix[i + 1, j + 1] = (matrix[i, j] + 
                                   prefix[i, j + 1] + 
                                   prefix[i + 1, j] - 
                                   prefix[i, j])
    
    # Step 2: Calculate sum for every possible 40x40 area
    area_sums = []
    max_i = m - area_size
    max_j = n - area_size
    
    for i in range(max_i + 1):
        for j in range(max_j + 1):
            # Calculate sum using prefix sums
            total = (prefix[i + area_size, j + area_size] - 
                    prefix[i, j + area_size] - 
                    prefix[i + area_size, j] + 
                    prefix[i, j])
            area_sums.append((total, i, j))
    
    # Step 3: Sort areas by sum in descending order
    area_sums.sort(reverse=True, key=lambda x: x[0])
    
    # Step 4: Greedily select non-overlapping areas
    def areas_overlap(area1: Tuple[int, int], area2: Tuple[int, int]) -> bool:
        """Check if two 40x40 areas overlap"""
        i1, j1 = area1
        i2, j2 = area2
        
        # Check if rectangles overlap
        return not (i1 + area_size <= i2 or i2 + area_size <= i1 or
                   j1 + area_size <= j2 or j2 + area_size <= j1)
    
    selected_areas = []
    
    for _, i, j in area_sums:
        if len(selected_areas) >= num_areas:
            break
            
        # Check if this area overlaps with any selected area
        overlaps = False
        for selected_i, selected_j in selected_areas:
            if areas_overlap((i, j), (selected_i, selected_j)):
                overlaps = True
                break
        
        if not overlaps:
            selected_areas.append((i, j))
    
    return selected_areas

# Alternative: More efficient approach using priority consideration
def find_max_areas_optimized(matrix: np.ndarray, area_size: int = 40, num_areas: int = 5) -> List[Tuple[int, int]]:
    """
    Optimized version that considers area values during selection.
    """
    m, n = matrix.shape
    
    # Precompute all area sums
    prefix = np.zeros((m + 1, n + 1))
    for i in range(m):
        for j in range(n):
            prefix[i + 1, j + 1] = (matrix[i, j] + 
                                   prefix[i, j + 1] + 
                                   prefix[i + 1, j] - 
                                   prefix[i, j])
    
    # Calculate all possible areas
    areas = []
    max_i = m - area_size
    max_j = n - area_size
    
    for i in range(max_i + 1):
        for j in range(max_j + 1):
            total = (prefix[i + area_size, j + area_size] - 
                    prefix[i, j + area_size] - 
                    prefix[i + area_size, j] + 
                    prefix[i, j])
            areas.append((total, i, j))
    
    # Sort by sum
    areas.sort(reverse=True, key=lambda x: x[0])
    
    def has_overlap(new_area, existing_areas):
        new_i, new_j = new_area
        for exist_i, exist_j in existing_areas:
            if (new_i < exist_i + area_size and new_i + area_size > exist_i and
                new_j < exist_j + area_size and new_j + area_size > exist_j):
                return True
        return False
    
    # Greedy selection with backtracking consideration
    selected = []
    
    # Try multiple candidates to get better global solution
    candidates_to_consider = min(1000, len(areas))  # Limit for performance
    
    for k in range(candidates_to_consider):
        if len(selected) >= num_areas:
            break
            
        current_selected = selected.copy()
        _, i, j = areas[k]
        
        # Try to add this candidate
        if not has_overlap((i, j), current_selected):
            current_selected.append((i, j))
            
            # Continue with remaining candidates
            for l in range(k + 1, candidates_to_consider):
                if len(current_selected) >= num_areas:
                    break
                _, next_i, next_j = areas[l]
                if not has_overlap((next_i, next_j), current_selected):
                    current_selected.append((next_i, next_j))
            
            # Keep the best selection found
            if len(current_selected) > len(selected):
                selected = current_selected
    
    return selected[:num_areas]

# Visualization function
def visualize_areas(matrix: np.ndarray, areas: List[Tuple[int, int]], area_size: int = 40):
    """
    Visualize the selected areas on the matrix.
    """
    import matplotlib.pyplot as plt
    
    plt.figure(figsize=(12, 8))
    plt.imshow(matrix, cmap='viridis', aspect='auto')
    plt.colorbar(label='Value')
    
    for idx, (i, j) in enumerate(areas):
        # Draw rectangle around the area
        rect = plt.Rectangle((j - 0.5, i - 0.5), area_size, area_size, 
                           linewidth=2, edgecolor='red', facecolor='none')
        plt.gca().add_patch(rect)
        plt.text(j + area_size/2, i + area_size/2, str(idx + 1), 
                color='white', fontsize=12, ha='center', va='center',
                bbox=dict(boxstyle="circle,pad=0.3", facecolor='red'))
    
    plt.title(f'Top {len(areas)} Non-overlapping {area_size}x{area_size} Areas')
    plt.xlabel('Column Index')
    plt.ylabel('Row Index')
    plt.show()

# Example usage and testing
def test_solution():
    # Create a sample matrix
    np.random.seed(42)
    matrix = np.random.rand(200, 300)  # 200x300 matrix
    
    # Add some high-value regions
    matrix[50:90, 100:140] += 5  # High sum area 1
    matrix[120:160, 50:90] += 4  # High sum area 2
    matrix[30:70, 200:240] += 3  # High sum area 3
    matrix[150:190, 150:190] += 2  # High sum area 4
    matrix[80:120, 250:290] += 1  # High sum area 5
    
    print("Matrix shape:", matrix.shape)
    
    # Find top 5 non-overlapping areas
    areas = find_max_non_overlapping_areas(matrix)
    
    print("\nTop 5 non-overlapping 40x40 areas (top-left indices):")
    for idx, (i, j) in enumerate(areas, 1):
        # Calculate actual sum for verification
        area_sum = np.sum(matrix[i:i+40, j:j+40])
        print(f"Area {idx}: position ({i}, {j}), sum = {area_sum:.2f}")
    
    # Visualize the results
    visualize_areas(matrix, areas)
    
    return areas

# Performance optimized version for very large matrices
def find_max_areas_large_matrix(matrix: np.ndarray, area_size: int = 40, num_areas: int = 5, stride: int = 5):
    """
    For very large matrices, use stride to reduce computation.
    """
    m, n = matrix.shape
    
    # Use stride to reduce number of positions to check
    positions = []
    for i in range(0, m - area_size + 1, stride):
        for j in range(0, n - area_size + 1, stride):
            positions.append((i, j))
    
    # Calculate sums only for strided positions
    area_sums = []
    for i, j in positions:
        total = np.sum(matrix[i:i+area_size, j:j+area_size])
        area_sums.append((total, i, j))
    
    area_sums.sort(reverse=True, key=lambda x: x[0])
    
    # Select non-overlapping areas (same logic as before)
    selected = []
    for _, i, j in area_sums:
        if len(selected) >= num_areas:
            break
        
        overlap = False
        for sel_i, sel_j in selected:
            if (i < sel_i + area_size and i + area_size > sel_i and
                j < sel_j + area_size and j + area_size > sel_j):
                overlap = True
                break
        
        if not overlap:
            selected.append((i, j))
    
    return selected

# Run the test
if __name__ == "__main__":
    result_areas = test_solution()
```

- Car Pooling Problem

```python
def car_pooling(trips, capacity, num_loc=1001):
    # Use prefix sum to track passenger changes
    locations = [0] * num_loc
    
    for passengers, start, end in trips:
        locations[start] += passengers
        locations[end] -= passengers
    
    current_passengers = 0
    for change in locations:
        current_passengers += change
        if current_passengers > capacity:
            return False
    
    return True

trips = [[2, 1, 5], [3, 3, 7]] # [[p, s, e], [p, s, e], ...]
capacity = 4
print(f"Can complete trips: {car_pooling(trips, capacity)}")  # False
```

```python
def max_subarray_sum_circular(arr):
    # Case 1: Normal maximum subarray (Kadane's)
    def kadane(nums):
        max_ending_here = max_so_far = nums[0]
        for num in nums[1:]:
            max_ending_here = max(num, max_ending_here + num)
            max_so_far = max(max_so_far, max_ending_here)
        return max_so_far
    
    # Case 2: Circular maximum (total - min_subarray)
    total_sum = sum(arr)
    min_ending_here = min_so_far = arr[0]
    for num in arr[1:]:
        min_ending_here = min(num, min_ending_here + num)
        min_so_far = min(min_so_far, min_ending_here)
    
    normal_max = kadane(arr)
    circular_max = total_sum - min_so_far
    
    # If all numbers are negative
    if circular_max == 0:
        return normal_max
    
    return max(normal_max, circular_max)

arr = [5, -3, 5]
print(f"Maximum circular subarray sum: {max_subarray_sum_circular(arr)}")  # 10
```

## <div class="alert alert-danger"> 2. Sliding Window

### Definition
- The Sliding Window technique is an algorithmic approach where we maintain a window (subarray/substring) that slides over the data structure (usually an array or string) to solve problems efficiently.
- Instead of processing each element individually, we process elements in chunks.

### Key Characteristics:
- Fixed-size window: Window size remains constant
- Variable-size window: Window size changes based on conditions
- Overlapping subproblems: Windows overlap with previous/next windows
- Efficient computation: Avoid redundant calculations by reusing previous results

### Common Use Cases
1. Fixed Size Window Problems
   - Maximum/minimum sum of subarray of size k
   - Average of subarrays of size k
   - First negative number in every window of size k

2. Variable Size Window Problems
   - Longest substring with k distinct characters
   - Minimum window substring
   - Longest substring without repeating characters
   - Subarray sum equals k

3. Other Applications
   - Pattern matching
   - Signal processing
   - Time series analysis
   - DNA sequence analysis

### Implementations

#### 1. Fixed Size Window Template

```python
def fixed_sliding_window(arr, k):
    """
    Template for fixed size sliding window problems
    """
    n = len(arr)
    if n < k:
        return None
    
    # Initialize first window
    window_sum = sum(arr[:k])
    result = [window_sum]
    
    # Slide the window
    for i in range(k, n):
        # Remove left element, add right element
        window_sum = window_sum - arr[i - k] + arr[i]
        result.append(window_sum)
    
    return result
```

- Maximum sum of subarray of size k

```python
# Example: Maximum sum of subarray of size k
def max_sum_subarray(arr, k):
    n = len(arr)
    if n < k:
        return None
    
    window_sum = sum(arr[:k])
    max_sum = window_sum
    
    for i in range(k, n):
        window_sum = window_sum - arr[i - k] + arr[i]
        max_sum = max(max_sum, window_sum)
    
    return max_sum

# Test
arr = [2, 1, 5, 1, 3, 2]
k = 3
print(f"Maximum sum of subarray of size {k}: {max_sum_subarray(arr, k)}")
# Output: Maximum sum of subarray of size 3: 9
```

### 2. Variable Size Window Template

```python
def variable_sliding_window(arr, target):
    """
    Template for variable size sliding window problems
    """
    left = 0
    current_sum = 0
    result = 0  # or float('inf') for minimization problems
    
    for right in range(len(arr)):
        # Add current element to window
        current_sum += arr[right]
        
        # Shrink window while condition is violated
        while current_sum > target and left <= right:
            current_sum -= arr[left]
            left += 1
        
        # Update result based on current window
        if current_sum == target:
            result = max(result, right - left + 1)
    
    return result
```

```python
# Example: Longest substring with at most k distinct characters
def longest_substring_k_distinct(s, k):
    if k == 0 or not s:
        return 0
    
    left = 0
    char_count = {}
    max_length = 0
    
    for right in range(len(s)):
        # Add right character to window
        char = s[right]
        char_count[char] = char_count.get(char, 0) + 1
        
        # Shrink window if we have more than k distinct characters
        while len(char_count) > k:
            left_char = s[left]
            char_count[left_char] -= 1
            if char_count[left_char] == 0:
                del char_count[left_char]
            left += 1
        
        # Update max length
        max_length = max(max_length, right - left + 1)
    
    return max_length

# Test
s = "araaci"
k = 2
print(f"Longest substring with {k} distinct characters: {longest_substring_k_distinct(s, k)}")
# Output: Longest substring with 2 distinct characters: 4
```

### 3. Complete Examples
- Example 1: Average of Subarrays of Size K

```python
def find_averages_of_subarrays(k, arr):
    result = []
    window_sum, window_start = 0.0, 0
    
    for window_end in range(len(arr)):
        window_sum += arr[window_end]  # add the next element
        
        # slide the window, no need to slide if we've not hit the required window size 'k'
        if window_end >= k - 1:
            result.append(window_sum / k)  # calculate the average
            window_sum -= arr[window_start]  # subtract the element going out
            window_start += 1  # slide the window ahead
    
    return result

# Test
arr = [1, 3, 2, 6, -1, 4, 1, 8, 2]
k = 5
print(f"Averages of subarrays of size {k}: {find_averages_of_subarrays(k, arr)}")
# Output: Averages of subarrays of size 5: [2.2, 2.8, 2.4, 3.6, 2.8]
```

#### Example 2: Smallest Subarray with Given Sum

```python
def smallest_subarray_with_given_sum(s, arr):
    window_sum = 0
    min_length = float('inf')
    window_start = 0
    
    for window_end in range(len(arr)):
        window_sum += arr[window_end]  # add the next element
        
        # shrink the window as small as possible until the window_sum is smaller than s
        while window_sum >= s:
            min_length = min(min_length, window_end - window_start + 1)
            window_sum -= arr[window_start]
            window_start += 1
    
    return min_length if min_length != float('inf') else 0

# Test
arr = [2, 1, 5, 2, 3, 2]
s = 7
print(f"Smallest subarray length with sum >= {s}: {smallest_subarray_with_given_sum(s, arr)}")
# Output: Smallest subarray length with sum >= 7: 2
```

#### Example 3: Longest Substring Without Repeating Characters

```python
def length_of_longest_substring(s):
    char_index = {}  # stores the most recent index of each character
    left = 0
    max_length = 0
    
    for right in range(len(s)):
        char = s[right]
        
        # if the character is already in the current window
        if char in char_index and char_index[char] >= left:
            # move the left pointer to the right of the previous occurrence
            left = char_index[char] + 1
        
        # update the character's most recent index
        char_index[char] = right
        
        # update the maximum length
        max_length = max(max_length, right - left + 1)
    
    return max_length

# Test
s = "abcabcbb"
print(f"Longest substring without repeating characters: {length_of_longest_substring(s)}")
# Output: Longest substring without repeating characters: 3
```

### Key Points to Remember
- When to use sliding window:
  - Problems involving subarrays/substrings
  - Need to find min/max/longest/shortest subarray
  - Array elements need to satisfy certain constraints

- Implementation tips:
  - Use two pointers (left and right) to represent window boundaries
  - Maintain a data structure (dictionary, set) to track window state
  - Expand the right pointer to include new elements
  - Shrink the left pointer when constraints are violated

- Common patterns:
  - For fixed size: Calculate first window, then slide with add/remove
  - For variable size: Expand right until condition fails, then shrink left

- The sliding window technique is powerful for optimizing solutions to array/string problems by reducing time complexity from O(n²) or O(n*k) to O(n).



## <div class="alert alert-warning"> 3. Two Pointers

- The two-pointer technique in Python involves using two indices, or "pointers," to traverse a data structure, most commonly an array or linked list, to solve problems more efficiently. This technique often reduces time complexity from O(n²) (brute force with nested loops) to O(n) (linear time). 
- Key Concepts and Applications:
  - Pointers at Opposite Ends (for Sorted Arrays):
    - Initialize one pointer at the beginning (left) and another at the end (right) of a sorted array.
    - Move the pointers towards each other based on a specific condition.
    - Example: Finding a pair of numbers that sum to a target in a sorted array. If the sum of elements at left and right is less than the target, increment left. If it's greater, decrement right. 

```python
def two_sum_sorted(nums, target):
    left, right = 0, len(nums) - 1
    while left < right:
        current_sum = nums[left] + nums[right]
        if current_sum == target:
            return [left, right]
        elif current_sum < target:
            left += 1
        else:
            right -= 1
    return None
```

### Definition
- The Two Sum problem is one of the most classic and frequently asked coding interview questions. Given an array of integers, find two numbers such that they add up to a specific target.

### Problem Statement
- Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

### Assumptions:
- Each input has exactly one solution
- You cannot use the same element twice
- Return the indices in any order

### Examples
- Example 1:
  - Input: nums = [2, 7, 11, 15], target = 9
  - Output: [0, 1]
  - Explanation: nums[0] + nums[1] = 2 + 7 = 9

- Example 2:
    - Input: nums = [3, 2, 4], target = 6
    - Output: [1, 2]
    - Explanation: nums[1] + nums[2] = 2 + 4 = 6

- Example 3:
    - Input: nums = [3, 3], target = 6
    - Output: [0, 1]

```python
# Brute Force
def two_sum_brute_force(nums, target):
    n = len(nums)
    for i in range(n):
        for j in range(i + 1, n):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []

# Time Complexity: O(n²)
# Space Complexity: O(1)
```

```python
# Hash Map (Optimal)
def two_sum_hashmap(nums, target):
    num_map = {}  # value -> index
    
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        num_map[num] = i
    
    return []

# Time Complexity: O(n)
# Space Complexity: O(n)
```

```python
# Two Pointers
def two_sum_sorted(nums, target):
    # If we need to return indices, we should store them before sorting
    indexed_nums = [(num, i) for i, num in enumerate(nums)]
    indexed_nums.sort()  # Sort by value
    
    left, right = 0, len(indexed_nums) - 1
    
    while left < right:
        current_sum = indexed_nums[left][0] + indexed_nums[right][0]
        
        if current_sum == target:
            return [indexed_nums[left][1], indexed_nums[right][1]]
        elif current_sum < target:
            left += 1
        else:
            right -= 1
    
    return []

# Time Complexity: O(n log n)
# Space Complexity: O(n)
```

```python
# Sorted Array

def two_sum_sorted_input(numbers, target):
    left, right = 0, len(numbers) - 1
    
    while left < right:
        current_sum = numbers[left] + numbers[right]
        
        if current_sum == target:
            return [left + 1, right + 1]  # 1-indexed
        elif current_sum < target:
            left += 1
        else:
            right -= 1
    
    return []

# Example: numbers = [2, 7, 11, 15], target = 9 → [1, 2]
```

```python
# Return All Two Sum Pairs
def two_sum_all_pairs(nums, target):
    num_map = {}
    result = []
    
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            for idx in num_map[complement]:
                result.append([idx, i])
        
        # Store all indices for duplicates
        if num not in num_map:
            num_map[num] = []
        num_map[num].append(i)
    
    return result

# Example: nums = [1, 3, 2, 2, 3, 1], target = 4
# Output: [[0, 1], [0, 4], [2, 3], [3, 2], [5, 1], [5, 4]]
```

### Definition
- The two pointers technique is an algorithmic approach where two pointers traverse a data structure (typically an array or linked list) to solve problems efficiently, often in linear time ***O(n)***.

### Key Characteristics
- Time Complexity: Typically O(n)

- 1. Two Sum in Sorted Array

```python
def two_sum_sorted(arr, target):
    left, right = 0, len(arr) - 1
    
    while left < right:
        current_sum = arr[left] + arr[right]
        
        if current_sum == target:
            return [left, right]
        elif current_sum < target:
            left += 1
        else:
            right -= 1
    
    return [-1, -1]

# Example
arr = [2, 4, 5, 7, 11, 15]
target = 9
print(two_sum_sorted(arr, target))  # Output: [0, 1]
```

- 2. Palindrome Check

```python
def is_palindrome(s):
    left, right = 0, len(s) - 1
    
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    
    return True

# Example
print(is_palindrome("racecar"))  # True
print(is_palindrome("hello"))    # False
```

- 3. Remove Duplicates from Sorted Array

```python
def remove_duplicates(nums):
    if not nums:
        return 0
    
    slow = 0
    
    for fast in range(1, len(nums)):
        if nums[fast] != nums[slow]:
            slow += 1
            nums[slow] = nums[fast]
    
    return slow + 1

# Example
nums = [1, 1, 2, 2, 2, 3, 4, 4, 5]
new_length = remove_duplicates(nums)
print(nums[:new_length])  # [1, 2, 3, 4, 5]
```

- 4. Container With Most Water

```python
def max_area(height):
    left, right = 0, len(height) - 1
    max_water = 0
    
    while left < right:
        # Calculate current area
        width = right - left
        current_height = min(height[left], height[right])
        current_area = width * current_height
        
        # Update max
        max_water = max(max_water, current_area)
        
        # Move pointer with smaller height
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1
    
    return max_water

# Example
height = [1, 8, 6, 2, 5, 4, 8, 3, 7]
print(max_area(height))  # Output: 49
```

- 5. Linked List Cycle Detection

```python
class ListNode:
    def __init__(self, val=0):
        self.val = val
        self.next = None

def has_cycle(head):
    if not head or not head.next:
        return False
    
    slow = head
    fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
        if slow == fast:
            return True
    
    return False
```

- 6. Three Sum
     - Definition: The Three Sum problem is a classic algorithmic challenge where you need to find all unique triplets in an array that sum up to a specific target (usually zero).

        - Problem Statement: Given an array of integers nums, find all unique triplets [nums[i], nums[j], nums[k]] such that:
        - i != j, i != k, and j != k
        - nums[i] + nums[j] + nums[k] == target (typically target = 0)
        - The solution set must not contain duplicate triplets

```python
def three_sum(nums, target=0):
    nums.sort()
    result = []
    
    for i in range(len(nums) - 2):
        # Skip duplicates
        if i > 0 and nums[i] == nums[i - 1]:
            continue
            
        left, right = i + 1, len(nums) - 1
        
        while left < right:
            current_sum = nums[i] + nums[left] + nums[right]
            
            if current_sum == target:
                result.append([nums[i], nums[left], nums[right]])
                
                # Skip duplicates
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                
                left += 1
                right -= 1
            elif current_sum < target:
                left += 1
            else:
                right -= 1
    
    return result

# Example
nums = [-1, 0, 1, 2, -1, -4]
print(three_sum(nums))  # Output: [[-1, -1, 2], [-1, 0, 1]]
```

### 1. The Big Idea: What is Dynamic Programming?

### 2. When to Use Dynamic Programming? (The Two Key Properties)
- Not every problem is suitable for DP. A problem must have two key properties:
#### Overlapping Subproblems:
- The problem can be broken down into smaller subproblems, and these subproblems are reused multiple times. The Fibonacci sequence is a perfect example. Recursive tree diagrams clearly show the same function calls being repeated.

#### Optimal Substructure:
- An optimal solution to the main problem can be constructed from optimal solutions of its subproblems. This is often phrased as the ability to write a recursive formula for the solution.
- Example (Shortest Path): If the shortest path from node A to node C goes through node B, then the path from A to B and from B to C must also be the shortest paths.


### 3. How to Solve a Problem with Dynamic Programming
- There are two fundamental approaches that implement the "store and reuse" idea. They are equivalent in their core logic but differ in implementation.

#### Approach 1: Top-Down (Memoization)
- This is essentially recursion with caching.
- How it works: You start with the original, large problem and recursively break it down. Before solving a subproblem, you check a cache (often called a "memo") to see if you've already solved it. If you have, you return the cached result. If you haven't, you compute it, store the result in the cache, and then return it.
- It's "lazy": You only solve subproblems as they are needed.
- Advantage: Often more intuitive to write, as it's very close to the recursive definition of the problem.
- Disadvantage: The recursion stack can be deep, leading to stack overflow errors for very large inputs.

#### Approach 2: Bottom-Up (Tabulation)
- This is the classic "table-filling" approach.
- How it works: You solve all the subproblems in a systematic order, starting from the smallest and simplest, and build up to the original problem. You store the results in a table (usually an array or matrix).
- It's "eager": You solve every possible subproblem, even if not all are needed for the final solution.
- Advantage: More efficient in terms of constant factors (no recursive function calls) and avoids stack overflow.
- Disadvantage: Can be less intuitive, and you might end up solving some unnecessary subproblems.

#### Comparison:

|Feature	|Top-Down (Memoization)	|Bottom-Up (Tabulation)|
|---|---|---|
|Approach	|Recursive	|Iterative|
|State	|The state of the subproblem is often passed as function parameters.	|The state is represented by indices in a table (e.g., dp[i]).|
|Order	|Solved on-demand as the recursion unfolds.	|Solved in a specific, predefined order (e.g., from dp[0] to dp[n]).|
|Speed	|Slower due to recursive calls.	|Faster, no function call overhead.|
|Stack |Overflow	Can happen with deep recursion.	|Not an issue.|

### 4. A Classic Example: The Fibonacci Sequence
- Let's implement both approaches for the Fibonacci numbers.
- Problem: Find the n-th Fibonacci number, where:
    - F(0) = 0, F(1) = 1, F(n) = F(n-1) + F(n-2) for n > 1.

#### a) Top-Down with Memoization (Python)

```python
def fib_memo(n, memo={}):
    # Check if we've already computed this subproblem
    if n in memo:
        return memo[n]
    
    # Base cases
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    # Recursive case: compute and store the result
    memo[n] = fib_memo(n-1, memo) + fib_memo(n-2, memo)
    return memo[n]

print(fib_memo(10))  # Output: 55
```

#### b) Bottom-Up with Tabulation (Python)

```python
def fib_tab(n):
    if n == 0:
        return 0
    
    # Create a DP table to store results
    dp = [0] * (n + 1)
    
    # Initialize base cases
    dp[0] = 0
    dp[1] = 1
    
    # Fill the table in a bottom-up order
    for i in range(2, n + 1):
        dp[i] = dp[i-1] + dp[i-2]
    
    return dp[n]

print(fib_tab(10))  # Output: 55
```

- Space Optimization: Notice for Fibonacci, you only need the last two values. You can optimize the space to O(1):

```python
def fib_optimized(n):
    if n == 0:
        return 0
    prev2, prev1 = 0, 1  # F(0), F(1)
    for i in range(2, n + 1):
        current = prev1 + prev2
        prev2 = prev1
        prev1 = current
    return prev1
```

### <font color='red'> 70. Climbing Stairs: </font>
- **Problem Statement**: You are climbing a staircase. It takes n steps to reach the top. Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?

#### Solution 1: Bottom-Up DP (Iterative)
- Time complexity = O(n)
- Space Complexity = O(n)

```python
def climbStairs(n: int) -> int:
    if n <= 2:
        return n
    
    # dp[i] represents number of ways to reach step i
    dp = [0] * (n + 1)
    dp[1] = 1  # Only one way: take 1 step
    dp[2] = 2  # Two ways: (1+1) or (2)
    
    for i in range(3, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]
    
    return dp[n]
```

- n=2: Just two distictive ways; in one stride equals to 2 steps or in two strides where one stride=two stairs.

```python
climbStairs(2)
```

- n=3: [one step at a time; one step and 2steps;  2 steps and one step]

```python
climbStairs(3)
```

#### Solution 2: Optimize space complexity:
  - Time complexity = O(n)
  - Space Complexity = O(1)

```python
def climbStairs(n: int) -> int:
    if n <= 2:
        return n
    
    # Only store the last two values
    prev2 = 1  # ways for n-2
    prev1 = 2  # ways for n-1
    
    for i in range(3, n + 1):
        current = prev2 + prev1
        prev2 = prev1
        prev1 = current
    
    return prev1

climbStairs(30)
```

#### Solution 3: Using Variables with Clear Names
  - Time complexity = O(n)
  - Space Complexity = O(1)

```python
def climbStairs(n: int) -> int:
    if n == 1:
        return 1
    
    one_step_before = 1  # ways to reach step 1
    two_steps_before = 1  # ways to reach step 0 (base case)
    current = 0
    
    for i in range(2, n + 1):
        current = one_step_before + two_steps_before
        two_steps_before = one_step_before
        one_step_before = current
    
    return current

climbStairs(30)
```

#### Solution 4: Recursive with Memoization (Top-Down DP)
  - Time complexity = O(n)
  - Space Complexity = O(n)

```python
def climbStairs(n: int) -> int:
    memo = {}
    
    def dfs(step):
        if step in memo:
            return memo[step]
        if step <= 2:
            return step
        
        memo[step] = dfs(step - 1) + dfs(step - 2)
        return memo[step]
    
    return dfs(n)

climbStairs(30)
```

### <font color='red'> 300. Longest Increasing Subsequence(LIS) </font>
- **Problem Statement**: Given an integer array nums, return the length of the longest strictly increasing subsequence.
- **Dynamic Programming Solution**:
  - Approach:
    - We can solve this using a DP array where dp[i] represents the length of the longest increasing subsequence ending at index i.
  - Algorithm:
    - Create a DP array of the same length as nums, initialized with 1 (each element is a subsequence of length 1)
    - For each position i, check all previous positions j (0 ≤ j < i)
    - If nums[i] > nums[j], update dp[i] = max(dp[i], dp[j] + 1)
    - Track the maximum value in the DP array

```python
def lengthOfLIS(nums):
    if not nums:
        return 0
    
    n = len(nums)
    dp = [1] * n  # Each element is at least a subsequence of length 1
    
    for i in range(1, n):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    
    return max(dp)

# Example usage
nums = [10, 9, 2, 5, 3, 7, 101, 18]
print(lengthOfLIS(nums))  # Output: 4
# Explanation: The longest increasing subsequence is [2, 3, 7, 101]
```

#### Step-by-Step Example
- Let's trace through the example nums = [10, 9, 2, 5, 3, 7, 101, 18]:
- Initialization:
    - nums: [10, 9, 2, 5, 3, 7, 101, 18]
    - dp:   [1,  1, 1, 1, 1, 1, 1,   1 ]
- Processing:
    - i=1: Compare with nums[0]=10 → 9 < 10 → no update
    - i=2: Compare with nums[0]=10, nums[1]=9 → 2 < both → no update
    - i=3: Compare with nums[2]=2 → 5 > 2 → dp[3] = max(1, 1+1) = 2
    - i=4: Compare with nums[2]=2 → 3 > 2 → dp[4] = max(1, 1+1) = 2
    - i=5:Compare with nums[2]=2 → 7 > 2 → dp[5] = max(1, 1+1) = 2
    - Compare with nums[3]=5 → 7 > 5 → dp[5] = max(2, 2+1) = 3
    - Compare with nums[4]=3 → 7 > 3 → dp[5] = max(3, 2+1) = 3
    - i=6:Compare with all previous elements, find max dp[j] where nums[j] < 101
    - Max is dp[5]=3 → dp[6] = 3+1 = 4
    - i=7: Compare with all previous, find max dp[j] where nums[j] < 18
    - Max is dp[5]=3 → dp[7] = 3+1 = 4
    - Final DP array: dp: [1, 1, 1, 2, 2, 3, 4, 4]

### <font color='red'> 312. Burst Balloons </font>

- ***Problem Statement***: You are given n balloons, indexed from 0 to n-1. Each balloon has a number, and when you burst balloon i, you get nums[i-1] * nums[i] * nums[i+1] coins. If i-1 or i+1 goes out of bounds, treat it as 1.
- ***Goal***: Find the maximum coins you can collect by bursting the balloons wisely.
- ***Key Insight***: The key insight is to think backwards - instead of thinking about which balloon to burst first, think about which balloon to burst last.
  - When you burst balloon i last in range [left, right], you get:
    - nums[left-1] * nums[i] * nums[right+1] coins, plus the maximum coins from bursting balloons in [left, i-1] and [i+1, right].

### DP Solution

```python
def maxCoins(nums):
    """
    Calculate maximum coins from bursting balloons
    
    Args:
        nums: List of balloon values
    
    Returns:
        Maximum coins that can be collected
    """
    # Add virtual balloons at boundaries with value 1
    n = len(nums)
    balloons = [1] + nums + [1]
    
    # dp[i][j] = max coins from bursting balloons from i to j (inclusive)
    dp = [[0] * (n + 2) for _ in range(n + 2)]
    
    # Iterate over lengths of subarrays
    for length in range(1, n + 1):  # length of subarray
        for left in range(1, n - length + 2):  # starting index
            right = left + length - 1  # ending index
            
            # Try every balloon in [left, right] as the last one to burst
            for last in range(left, right + 1):
                # Coins from bursting last balloon
                coins = balloons[left - 1] * balloons[last] * balloons[right + 1]
                
                # Coins from left and right subarrays
                coins += dp[left][last - 1] + dp[last + 1][right]
                
                # Update maximum
                dp[left][right] = max(dp[left][right], coins)
    
    return dp[1][n]

# Example usage
nums = [3, 1, 5, 8]
print(f"Maximum coins: {maxCoins(nums)}")  # Output: 167
```

### Step-by-Step Explanation
1. State Definition:
   - dp[i][j] = maximum coins we can collect by bursting balloons from index i to j (inclusive)
   - We use 1-based indexing for convenience with boundary handling

2. Base Case
   - When i > j, dp[i][j] = 0 (no balloons to burst)
   - When i == j, it's a single balloon: dp[i][i] = nums[i-1] * nums[i] * nums[i+1]

3. Recurrence Relation
   - For range [i, j], if we burst balloon k last:

```python
p[i][j] = max(
    balloons[i-1] * balloons[k] * balloons[j+1] + dp[i][k-1] + dp[k+1][j]
    for all k in [i, j]
)
```

### 4. Filling Order
- We fill the DP table by increasing length of subarrays:
- First handle all subarrays of length 1
- Then length 2, 3, ..., up to length n
- Example Walkthrough
  - For nums = [3, 1, 5, 8]:
  - Array with boundaries: [1, 3, 1, 5, 8, 1]

- Fill DP table:
  - Length 1: dp[1][1] = 1*3*1 = 3, etc.
  - Length 2: dp[1][2] = max(1*3*5 + dp[2][2], 1*1*5 + dp[1][1]), etc.
  - Continue until dp[1][4] gives the answer

- Alternative Implementation (More Readable)

```python
def maxCoins(nums):
    if not nums:
        return 0
    
    n = len(nums)
    # Add boundary balloons
    balloons = [1] + nums + [1]
    
    # dp[i][j] = max coins from bursting balloons i through j
    dp = [[0] * (n + 2) for _ in range(n + 2)]
    
    for length in range(1, n + 1):
        for i in range(1, n - length + 2):
            j = i + length - 1
            for k in range(i, j + 1):
                # k is the last balloon to burst in range [i, j]
                left_value = balloons[i - 1]
                right_value = balloons[j + 1]
                current_coin = left_value * balloons[k] * right_value
                
                total_coin = (dp[i][k - 1] + 
                            current_coin + 
                            dp[k + 1][j])
                
                dp[i][j] = max(dp[i][j], total_coin)
    
    return dp[1][n]
```

### Complexity Analysis
- Time Complexity: O(n³) - Three nested loops
- Space Complexity: O(n²) - For the DP table

### Key Takeaways
- Think backwards: Consider which balloon to burst last rather than first
- Define boundaries: Add virtual balloons at boundaries to handle edge cases
- Build from smaller to larger: Solve for smaller subarrays first
- State definition matters: dp[i][j] represents the maximum coins from bursting balloons in range [i, j]
- This approach ensures we consider all possible bursting orders and find the optimal solution efficiently.

### <font color='red'> 322. Coin Change </font>

- ***Problem Statement***: Given a set of coin denominations and a target amount, find the minimum number of coins needed to make that amount, or return -1 if it's impossible.

### Solution 1: Top-Down DP (Memoization)

```python
def coinChange(coins, amount):
    def dfs(remaining, memo):
        if remaining == 0:
            return 0
        if remaining < 0:
            return float('inf')
        if remaining in memo:
            return memo[remaining]
        
        min_coins = float('inf')
        for coin in coins:
            result = dfs(remaining - coin, memo)
            if result != float('inf'):
                min_coins = min(min_coins, result + 1)
        
        memo[remaining] = min_coins
        return min_coins
    
    result = dfs(amount, {})
    return result if result != float('inf') else -1

# Example usage
coins = [1, 2, 5]
amount = 11
print(coinChange(coins, amount))  # Output: 3 (5 + 5 + 1)
```

### Solution 2: Bottom-Up DP (Tabulation)

```python
def coinChange(coins, amount):
    # Initialize DP array with infinity (impossible amounts)
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # Base case: 0 coins needed for amount 0
    
    for i in range(1, amount + 1):
        for coin in coins:
            if i - coin >= 0:
                dp[i] = min(dp[i], dp[i - coin] + 1)
    
    return dp[amount] if dp[amount] != float('inf') else -1

# Example usage
coins = [1, 5, 10, 25] # Penny, Nickel, Dime, & Quater
amount = 110021
print(coinChange(coins, amount))  # Output: 3
```

### Solution 3: Space-Optimized Version

```python
def coinChange(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    
    return dp[amount] if dp[amount] != float('inf') else -1

# Example usage
coins = [1, 2, 5]
amount = 11
print(coinChange(coins, amount))  # Output: 3
```

### Solution 4: With Coin Combination Tracking

```python
def coinChangeWithCombination(coins, amount):
    dp = [float('inf')] * (amount + 1)
    coin_used = [-1] * (amount + 1)  # Track which coin was used
    dp[0] = 0
    
    for i in range(1, amount + 1):
        for coin in coins:
            if i - coin >= 0 and dp[i - coin] + 1 < dp[i]:
                dp[i] = dp[i - coin] + 1
                coin_used[i] = coin
    
    if dp[amount] == float('inf'):
        return -1, []
    
    # Reconstruct the coin combination
    combination = []
    current = amount
    while current > 0:
        coin = coin_used[current]
        combination.append(coin)
        current -= coin
    
    return dp[amount], combination

# Example usage
coins = [1, 2, 5]
amount = 11
min_coins, combination = coinChangeWithCombination(coins, amount)
print(f"Minimum coins: {min_coins}")  # Output: 3
print(f"Combination: {combination}")  # Output: [1, 5, 5] or similar
```

### Solution 5: Count Number of Ways (Different Problem)
- If you want to count the number of ways to make the amount (not minimum coins):

```python
def countWays(coins, amount):
    dp = [0] * (amount + 1)
    dp[0] = 1  # One way to make amount 0: use no coins
    
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] += dp[i - coin]
    
    return dp[amount]

# Example usage
coins = [1, 2, 5]
amount = 5
print(countWays(coins, amount))  # Output: 4 (1+1+1+1+1, 1+1+1+2, 1+2+2, 5)
```

### Complexity Analysis
- Time Complexity: O(amount × number_of_coins)
- Space Complexity: O(amount) for the DP array

### <font color='red'> 416. Partition Equal Subset Sum </font>

- ***Problem Statement***: Given a non-empty array of positive integers, determine if the array can be partitioned into two subsets such that the sum of elements in both subsets is equal.

            Input: [1, 5, 11, 5]
            Output: true
            Explanation: The array can be partitioned as [1, 5, 5] and [11]

- ***Key Insight****: If we can partition the array into two subsets with equal sum, then:
  - The total sum must be even
  - Each subset must sum to total_sum / 2
  - This becomes a subset sum problem where we check if there's a subset that sums to target = total_sum / 2

### Solution Approaches: 

#### Approach 1: Dynamic Programming (Recommended)
- Time Complexity: O(n × target)
- Space Complexity: O(target)

```python
def canPartition(nums):
    total_sum = sum(nums)
    
    # If total sum is odd, cannot partition
    if total_sum % 2 != 0:
        return False
    
    target = total_sum // 2
    n = len(nums)
    
    # DP table: dp[i] means whether sum i can be achieved
    dp = [False] * (target + 1)
    dp[0] = True  # Empty subset sums to 0
    
    for num in nums:
        # Iterate backwards to avoid reusing the same element
        for j in range(target, num - 1, -1):
            if dp[j - num]:
                dp[j] = True
    
    return dp[target]


nums=[1, 5, 11, 5]
canPartition(nums)
```

#### Approach 2: 2D Dynamic Programming (More Intuitive)
- Time Complexity: O(n × target)
- Space Complexity: O(n × target)

```python
def canPartition(nums):
    total_sum = sum(nums)
    if total_sum % 2 != 0:
        return False
    
    target = total_sum // 2
    n = len(nums)
    
    # dp[i][j] = can we achieve sum j using first i elements
    dp = [[False] * (target + 1) for _ in range(n + 1)]
    
    # Base case: sum 0 can always be achieved with empty subset
    for i in range(n + 1):
        dp[i][0] = True
    
    for i in range(1, n + 1):
        for j in range(1, target + 1):
            if j < nums[i - 1]:
                # Cannot include current element
                dp[i][j] = dp[i - 1][j]
            else:
                # Either exclude or include current element
                dp[i][j] = dp[i - 1][j] or dp[i - 1][j - nums[i - 1]]
    
    return dp[n][target]


nums=[1, 5, 11, 5]
canPartition(nums)
```

#### Step-by-Step Example
- Let's trace through the example: [1, 5, 11, 5]
- Total sum = 1 + 5 + 11 + 5 = 22
- Target = 22 / 2 = 11
#### DP table evolution:
        
        Initial: dp[0] = True, others = False
        
        Process num = 1:
        j from 11 to 1: dp[1] = True
        
        Process num = 5:
        j from 11 to 5:
        - j=11: dp[11] = dp[11] or dp[6] = False or False = False
        - j=10: dp[10] = dp[10] or dp[5] = False or False = False  
        - j=9: dp[9] = dp[9] or dp[4] = False or False = False
        - j=8: dp[8] = dp[8] or dp[3] = False or False = False
        - j=7: dp[7] = dp[7] or dp[2] = False or False = False
        - j=6: dp[6] = dp[6] or dp[1] = False or True = True
        - j=5: dp[5] = True
        
        Current dp: [T, T, F, F, F, T, T, F, F, F, F, F]
        
        Process num = 11:
        j from 11 to 11:
        - j=11: dp[11] = dp[11] or dp[0] = False or True = True
        
        Found target 11! Return True

#### Optimization Notes
- Early termination: If any element > target, return False immediately
- Sorting: Sort in descending order might help find solution faster
- Space optimization: Use 1D DP array instead of 2D

#### Optimized Solution

```python
def canPartition(nums):
    total_sum = sum(nums)
    if total_sum % 2 != 0:
        return False
    
    target = total_sum // 2
    
    # Early termination if any number is larger than target
    if any(num > target for num in nums):
        return False
    
    dp = [False] * (target + 1)
    dp[0] = True
    
    # Sort in descending order for potential optimization
    nums.sort(reverse=True)
    
    for num in nums:
        for j in range(target, num - 1, -1):
            if dp[target]:  # Early termination if target found
                return True
            if dp[j - num]:
                dp[j] = True
    
    return dp[target]
```

- This problem demonstrates a classic application of the subset sum problem using dynamic programming, which is efficient for moderate input sizes.

### <font color='red'> 1143. Longest Common Subsequence </font>

- ***Problem Statement***: Given two strings text1 and text2, return their longest common subsequence. If there is no common subsequence, return 0.

#### Basic LCS Implementation

```python
def longest_common_subsequence(text1, text2):
    """
    Find the length of the longest common subsequence between two strings.
    
    Args:
        text1 (str): First string
        text2 (str): Second string
    
    Returns:
        int: Length of the longest common subsequence
    """
    m, n = len(text1), len(text2)
    
    # Create a DP table with (m+1) x (n+1) dimensions
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    # Fill the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if text1[i - 1] == text2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    
    return dp[m][n]
```

#### LCS with Sequence Reconstruction

```python
def longest_common_subsequence_with_sequence(text1, text2):
    """
    Find the longest common subsequence and return both length and the actual sequence.
    
    Args:
        text1 (str): First string
        text2 (str): Second string
    
    Returns:
        tuple: (length, subsequence)
    """
    m, n = len(text1), len(text2)
    
    # Create DP table
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    # Fill DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if text1[i - 1] == text2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    
    # Reconstruct the subsequence
    lcs = []
    i, j = m, n
    
    while i > 0 and j > 0:
        if text1[i - 1] == text2[j - 1]:
            lcs.append(text1[i - 1])
            i -= 1
            j -= 1
        elif dp[i - 1][j] > dp[i][j - 1]:
            i -= 1
        else:
            j -= 1
    
    # Reverse to get the correct order
    lcs = ''.join(reversed(lcs))
    
    return dp[m][n], lcs
```

#### Space-Optimized Version

```python
def longest_common_subsequence_optimized(text1, text2):
    """
    Space-optimized version using O(min(m, n)) space.
    
    Args:
        text1 (str): First string
        text2 (str): Second string
    
    Returns:
        int: Length of the longest common subsequence
    """
    # Ensure text1 is the shorter string for space optimization
    if len(text1) > len(text2):
        text1, text2 = text2, text1
    
    m, n = len(text1), len(text2)
    
    # Use only two rows instead of full matrix
    prev = [0] * (n + 1)
    curr = [0] * (n + 1)
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if text1[i - 1] == text2[j - 1]:
                curr[j] = prev[j - 1] + 1
            else:
                curr[j] = max(prev[j], curr[j - 1])
        
        # Swap rows for next iteration
        prev, curr = curr, prev
    
    return prev[n]
```

#### Complete Example with Visualization

```python
def print_dp_table(text1, text2, dp):
    """Helper function to print the DP table."""
    m, n = len(text1), len(text2)
    
    print("DP Table:")
    print("    ", end="")
    for j in range(n + 1):
        if j == 0:
            print("  ", end="")
        else:
            print(f" {text2[j-1]}", end="")
    print()
    
    for i in range(m + 1):
        if i == 0:
            print("  ", end="")
        else:
            print(f"{text1[i-1]} ", end="")
        
        for j in range(n + 1):
            print(f"{dp[i][j]:2}", end=" ")
        print()
    print()

def lcs_complete(text1, text2):
    """
    Complete LCS implementation with visualization.
    
    Args:
        text1 (str): First string
        text2 (str): Second string
    
    Returns:
        dict: Complete results including length, sequence, and DP table
    """
    m, n = len(text1), len(text2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    # Fill DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if text1[i - 1] == text2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    
    # Reconstruct sequence
    lcs_sequence = []
    i, j = m, n
    
    while i > 0 and j > 0:
        if text1[i - 1] == text2[j - 1]:
            lcs_sequence.append(text1[i - 1])
            i -= 1
            j -= 1
        elif dp[i - 1][j] > dp[i][j - 1]:
            i -= 1
        else:
            j -= 1
    
    lcs_sequence = ''.join(reversed(lcs_sequence))
    
    return {
        'length': dp[m][n],
        'sequence': lcs_sequence,
        'dp_table': dp
    }
```

<font color='blue'> ***Climbing Stairs***</font>: You are climbing a staircase. It takes n steps to reach the top. Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?

- Naive Recursive Solution: $TC=O(2^n) \& \ SC=O(n)$

```python
def climb_stairs_naive(n):
    """
    Naive recursive solution (exponential time)
    Time: O(2^n), Space: O(n) for call stack
    """
    if n <= 1:
        return 1
    
    # Ways to reach n = ways to reach (n-1) + ways to reach (n-2)
    return climb_stairs_naive(n - 1) + climb_stairs_naive(n - 2)


climb_stairs_naive(3), climb_stairs_naive(5)
```

- Optimal DP Solution: Approach 1: Bottom-up DP (Tabulation), $TC=O(n) \& \ SC=O(n)$

```python
def climb_stairs_dp(n):
    """
    Bottom-up DP solution
    Time: O(n), Space: O(n)
    """
    if n <= 1:
        return 1
    
    # dp[i] represents number of ways to reach step i
    dp = [0] * (n + 1)
    dp[0] = 1  # Base case: 1 way to be at step 0
    dp[1] = 1  # Base case: 1 way to reach step 1
    
    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]
    
    return dp[n]

climb_stairs_dp(3), climb_stairs_dp(5)
```

- Optimal DP Solution: Approach 2: Space Optimized DP, $TC=O(n)\ \& \ SC=O(1)$

```python
def climb_stairs_optimized(n):
    """
    Space-optimized DP solution
    Time: O(n), Space: O(1)
    """
    if n <= 1:
        return 1
    
    # Only need to track the last two values
    prev2 = 1  # ways to reach step i-2
    prev1 = 1  # ways to reach step i-1
    
    for i in range(2, n + 1):
        current = prev1 + prev2
        prev2 = prev1
        prev1 = current
    
    return prev1

climb_stairs_optimized(3), climb_stairs_optimized(5)
```

```python
def check_anagrams_list(s1, s2):
    if (len(s1) != len(s2)) or (len(list(set(s1)-set(s2)))!=0):
        return False
        
    if not s1 and not s2:
        return True      
    
    if all([s1.count(ch)==s2.count(ch) for ch in s1]):
        return True
    else:
        return False
            

s1='danger'
s2='garden'
check_anagrams_list(s1, s2)    
```

2. Given a sorted array of integers ***arr*** and an integer ***target***, find the index of the first and last positions of the ***target*** in the arr. If the integer can't be found in ***arr***, return ***[-1, -1]***.

```python
def get_first_and_last_indices_target(arr, target):
    indices = [-1, -1]
    try:
        # WC = O(n), traverse via ll e/ts
        # BC = O(1), target is the first e/t
        indices[0]=arr.index(target)
    except:
        pass
        
    try:
        indices[1]=len(arr)-arr[::-1].index(target)-1
    except:
        pass

    return indices
        
```

```python
def run_length_encode(sequence):
    encoded_sequence = []
    if not sequence:
        return ""

    count = 1
    current_char = sequence[0]

    for i in range(1, len(sequence)):
        if sequence[i] == current_char:
            count += 1
        else:
            encoded_sequence.append(f"{current_char}{count}")
            current_char = sequence[i]
            count = 1
    encoded_sequence.append(f"{current_char}{count}")
    return "".join(encoded_sequence)

original_sequence = "AAAABBB1111CCCCDDAAA11"
encoded_result = run_length_encode(original_sequence)
print(encoded_result)
```

```python
def create_prefix_sum(arr):
    prefix_sum = [0] * (len(arr) + 1)  # Initialize with 0 at index 0 for easier calculation
    for i in range(len(arr)):
        prefix_sum[i+1] = prefix_sum[i] + arr[i] # For numerical arrays
    return prefix_sum

def create_vowel_prefix_sum(s):
    vowels = "aeiouAEIOU"
    prefix_vowel_count = [0] * (len(s) + 1)
    for i in range(len(s)):
        prefix_vowel_count[i+1] = prefix_vowel_count[i] + (1 if s[i] in vowels else 0)
    return prefix_vowel_count
```

- Calculate Substring Property using Prefix Sums:
  - To find the property of a substring from index left to right (inclusive):
  - The value is prefix_sum[right + 1] - prefix_sum[left].

```python
# Example for numerical array
numbers = [3, 4, 1, 7, 9, 1]
prefix_sums = create_prefix_sum(numbers)
# Get sum of substring from index 2 to 4 (inclusive: 1, 7, 9)
substring_sum = prefix_sums[4 + 1] - prefix_sums[2] 
print(f"Sum of substring: {substring_sum}") # Output: 17

# Example for string (vowel count)
word = "picture"
prefix_vowels = create_vowel_prefix_sum(word)
# Get vowel count in substring from index 0 to 3 (inclusive: pict)
vowel_count = prefix_vowels[3 + 1] - prefix_vowels[0]
print(f"Vowel count in substring: {vowel_count}") # Output: 1
```
